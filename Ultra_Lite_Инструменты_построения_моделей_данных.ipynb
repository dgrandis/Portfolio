# ERD диаграмма - это диаграмма сущность свзяь

Для начала нужно определиться что имеют в своем составе сущности.

## **Фильм:**
  - Название
  - Жанр
  - Рейтинг
  - Продолжительность
  - Год выхода
  - Бюджет
  - Постер
  - Обзор
  - Кассовые сборы

# Каждый фильм имеет **персонажа**:
  - Имя
  - Актер

## Каждый фильм имеет съемочный **состав**:
 - Имя
 - Роль в съемке

# Съмочный состав имеет **персонал**: 
  - Имя
  - Род занятий

# **Так же Актеры**:
- Фамилия
- Имя
- Род занятий
- Фильмография
- Биография
- Дата рождения
- Место рождения
- Пол
- Фотографии
- Награды

## **Сериал:**
  - Название
  - Жанр
  - Сезоны 
  - Рейтинг
  - Продолжительность
  - Год выхода
  - Бюджет
  - Постер
  - Обзор

Так же сериал имеет сезоны:
  - Эпизод
  - Дата выхода

И каждый сезон имеет эпизоды:
 - Название
 - Дата выхода
 - Продолжительность

Попробуем изобрать все это на диаграмме **ERD**
