Задача 3

Дискретная случайная величина задана своим законом распределения:

Найти её математическое ожидание, дисперсию и среднеквадратическое отклонение.

```
W = {-2:0.4, 0:0.1, 3:0.3, 7:0.2}                       # распределение случайной велечины

import numpy as np                                      # Импортируем бибилиотеку numpy
E = sum([W[a]*a for a in W.keys()])                     # Математическое ожидание (сумма произведения случайной величины и вероятности)
E_x2 = sum([(key**2) * W[key] for key in W.keys()])     # Сумма произведения квадратов случайной величины и вероятности
Disp = round(E_x2 - (E**2),4)                           # Дисперсия
SKO = round(np.sqrt(E_x2 - (E**2)),3)                   # Среднеквадратичное отклонение

print('Мат.ожидание =', E)
print('Дисперсия =', Disp)
print('Среднеквадратичное отклонение =',SKO)
```
