### Задача 3

Используя соответствующий встроенный метод, дропните, т. е. удалите из нашей таблички все столбцы, у которых количество пропущенных значений больше 200.

* Найдите все такие столбцы (метод .isna() + подвыборка с условием).
* Определите их названия.
* Дропните их (используйте метод .drop(имена столбцов)).
* Результат сохраните в новой табличке data_1, исходную менять не нужно.

Могут быть полезны следующие параметры и методы:

*   columns - названия столбцов датафрейма
*   df[условие] - подвыборка элементов по условию
*   .drop - удаление строк или столбцов (в зависимости от параметра axis)



**Решение:**

```
mask = train.columns[train.isna().sum() > 200]                            # Делаем маску, которая удовлетворяет данные условия
#temp = (train.columns[mask])                                             # Записываем выборку столбцов в переменную temp (таких 6 столбцов)
print ( "Данные столбцы нужно удалить:\n\n",mask)                         # Вывод на экран названий этих столбцов (Просто посмотреть)

data_1 = train.drop(mask, axis = 1)                                       # Удаляем столбцы по маске и результат присваеваем переменной data_1

print ("\nРазмерность таблицы: ",data_1.shape)                            # Вывод размерности. Видим что количество столбцов стало 74 (уменьшилось на 6)
print (train.shape)
data_1.head()                                                             # Выводим первые 5 строк и видим, что нет данных столбцов (temp)

#data_1 = train.drop(train.columns[mask], axis = 'columns')               # Можно обойтись без переменной temp 



```
