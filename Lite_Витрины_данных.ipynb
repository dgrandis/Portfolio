# Задание **Lite**

Выполните задание ultra light.

Витрина должна удовлетворять следующим требованиям:

1. Содержать агрегированные показатели
2. Иметь цель (пропишите, какую задачу решает витрина)

Данную работу можно посомтреть на гугл документах по ссылке: https://docs.google.com/document/d/1D1U2WjRjNqNurzwxvzVa8Q6VUYRcrygPyQh7pvbKsqE/edit?usp=sharing

# 1. Содержать агрегированные показатели

Витрина для **продаж**: цена товара=цена поставщика+свои затраты+сумма для прибыли

Витрина для **производства**: сколько  еще надо произвести = необходимое кол-во для продаж - остатки готовой продукции 


Витрина для **закупок**: сколько надо еще произвести = необходимое кол-во для производства - остатки сырья



# 2 .Иметь цель (пропишите, какую задачу решает витрина)

Витрина для **продаж**:

*Цель*: рассчитать цену продажи с учетом компенсации затрат и получения прибыли

Витрина для **производства**: 

*Цель*: исключить возможность перепроизводства во избежании избытка на складе


Витрина для **закупок**:

*Цель*: исключить возможность излишней закупки во избежании избытка на складе.
