## **Задача 1**

Подключитесь к Базе данных и выведите из таблицы pg_attribute  только те записи в которых поле attbyval = True

Используйте оператор WHERE

Сначала подключимся к Базе данных

image.png

```python
# Далее нужно выбрать все столбцы из таблицы pg_attribute с помощью *
# И указать с помощью оператора WHERE задаем нужное нам условие

SELECT * from pg_attribute WHERE attbyval = True
```

Как видно вывелись только те строки из таблицы значение коотрые в столбце **attbyval** равняется **True**

## **Задача 2**

Подключитесь к Базе данных и выведите из таблицы pg_cast только записи где значения в поле oid больше 11330, но меньше 11337

```python
# Так же с помощью SELECT выбираем все столбцы из таблицы pg_cast
# И с оператора WHERE указываем нужный диапазон поля от 11330 до 11337

SELECT * from pg_cast WHERE oid > 11330 and oid < 11337
```
