**Задача 1**

Напишите функцию, которая будет выводить количество букв в слове. Проверьте при помощи этой функции, сколько букв в слове восьмидесятивосьмимиллиметровое.

**Решение:**

```
# Просим пользователя ввести любое слово, у которого хотим вывести кол-во букв
# И записываем в глобальную переменную word
word = input('Введите слово:')

# Создаем функцию которая будет считать кол-во букв в слове, который ввел пользователь
def number_of_letters():
  i = len(word)
  print (f'В слове {word} - {i} букв(а)')

# Вызываем данную функцию
number_of_letters()
```

```
# Можно сделать другим способом, без глобальной переменной, а передать нужное слово при вызове функции

def number_of_letters(word):
  i = len(word)
  return i

number_of_letters('восьмидесятивосьмимиллиметровое')
```

**Задача 2**

Напишите функцию, которая выводит таблицу умножения (от 1 до 9) для указанного числа в следующем формате:

1 * 3 = 3

2 * 3 = 6

3 * 3 = 9 

...

9 * 3 = 27

**Решение:**

```
# Так как нужно вывести конкретные значения, то можно их определить внутри функции

# Определяем функцию, которая будет выводить строки
def mult_on_3 ():
  for i in range(9):                           # Используем цикл for 
    print (i+1,'*', '3 = ', (i+1)*3)

# Вызываем функцию
mult_on_3()
  
```

```
# Так же можно создать функцию которая выводит таблицу умножения для введенного числа

x = int(input('Введите число:'))

def mult ():
  for i in range(9):
    print(f'{i+1} * {x}  = ',(i+1)*x )

# Вызываем функцию    
mult ()
```

**Задача 3**

Напишите функцию, которая выводит сумму, разность и произведение двух, поданных на вход чисел.


**Решение:**

```
def calc (int_1, int_2):
  print (f'{int_1} + {int_2} = {int_1 + int_2}')
  print (f'{int_1} - {int_2} = {int_1 - int_2}')
  print (f'{int_1} * {int_2} = {int_1 * int_2}')

# Вызываем функцию и в качестве аргументов передаем ей два числа, у которых хотим узнать сумму, разность и произведение
calc(4,5)
```

**Задача 4**

Импортируйте функцию sin из модуля math с присвоением ей псевдонима sinus. Посчитайте при помощи импортированной функции, чему будет равен синус единицы.



**Решение:**

```
# Из библиотеки math импортируем функцию sin и присваеваем ей имя sinus
from math import sin as sinus

# Создаем функцию которая принимает на вход число и выдает значение синуса данного числа
def sin(x):
  i = sinus(x)
  return i

# Вызываем функцию на вход передаем единицу
sin(1)
```

**Задача 5**

Напишите лямбда-функцию, которая считает куб числа.



**Решение:**

```
cube = lambda x: x**3
cube(3)
```

```
# Так же можно запросить число у пользователя
x = int(input("Введите число:"))
number_cube = lambda x: x**3
print (f'{x} в кубе равно: {number_cube(x)}')
```
