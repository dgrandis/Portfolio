# Выгрузите SQL код для созданной схемы

CREATE TABLE `Сотрудники` (

	`ID сотрудника` INT NOT NULL,
	`Фамилия` VARCHAR(255) NOT NULL,
	`Имя` VARCHAR(255) NOT NULL,
	`Отчество` VARCHAR(255) NOT NULL,
	`Должность` VARCHAR(255) NOT NULL,
	`Дата рождения` VARCHAR(255) NOT NULL,
	`Пол` BINARY NOT NULL,
	`Адрес` VARCHAR(255) NOT NULL,
	`Телефон` varchar(12) NOT NULL,
	PRIMARY KEY (`ID сотрудника`)
);

CREATE TABLE `Заказы` (

	`ID заказа` BINARY NOT NULL,
	`Дата заказа` DATETIME NOT NULL,
	`Количество шт` INT NOT NULL,
	`Сумма заказа` FLOAT NOT NULL,
	`Способ доставки` VARCHAR(255) NOT NULL,
	PRIMARY KEY (`ID заказа`)
);

CREATE TABLE `Клиенты` (

	`ID клиента` BINARY NOT NULL,
	`Фамилия` VARCHAR(255) NOT NULL,
	`Имя` VARCHAR(255) NOT NULL,
	`Отчество` VARCHAR(255) NOT NULL,
	`Адрес` VARCHAR(255) NOT NULL,
	`Телефон` varchar(12) NOT NULL,
	PRIMARY KEY (`ID клиента`)
);

CREATE TABLE `Покупки` (

	`ID покупки` INT NOT NULL,
	`ID клиента` INT NOT NULL,
	`Дать покупки` DATETIME NOT NULL,
	`Цена` FLOAT NOT NULL,
	`Количество` INT NOT NULL,
	`НДС` FLOAT NOT NULL,
	`Сумма НДС` FLOAT NOT NULL,
	PRIMARY KEY (`ID покупки`)
);

CREATE TABLE `Книги` (

	`ID товара` INT NOT NULL,
	`Наименования` VARCHAR(255) NOT NULL,
	`Автор` VARCHAR(255) NOT NULL,
	`Жанр` VARCHAR(255) NOT NULL,
	`Год издания` DATETIME NOT NULL,
	`Кол-во страниц` INT(255) NOT NULL,
	`Цена` FLOAT NOT NULL,
	`Количество` INT NOT NULL,
	PRIMARY KEY (`ID товара`)
);

CREATE TABLE `Поставщики` (

	`ID поставщика` INT NOT NULL,
	`ФИО` VARCHAR(255) NOT NULL,
	`Наименование` VARCHAR(255) NOT NULL,
	`Юридический адрес` VARCHAR(255) NOT NULL,
	`Телефон` varchar(12) NOT NULL,
	PRIMARY KEY (`ID поставщика`)
);

CREATE TABLE `Склад` (
  
	`ID поставки` INT NOT NULL AUTO_INCREMENT,
	`ID товара` INT NOT NULL AUTO_INCREMENT,
	`ID издательства` INT NOT NULL AUTO_INCREMENT,
	`ID поставщика` INT NOT NULL AUTO_INCREMENT,
	`№ товарной накладной` VARCHAR(255) NOT NULL AUTO_INCREMENT,
	`Дата поставки` DATETIME NOT NULL AUTO_INCREMENT,
	`Цена` FLOAT NOT NULL AUTO_INCREMENT,
	`Количество` INT NOT NULL AUTO_INCREMENT,
	`Остаток` INT NOT NULL AUTO_INCREMENT,
	PRIMARY KEY (`ID поставки`)
);
