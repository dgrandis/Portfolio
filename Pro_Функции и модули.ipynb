**Задача 1**

Напишите функцию, которая принимает на вход строку и считает количество букв в  верхнем и нижнем регистрах. 
Пример: 'На берегу Амазонки сидел Фернандо'.
Что должно быть на выходе?

Количество букв в верхнем регистре: 3

Количество букв в нижнем регистре: 26

**Решение:**

````
def up_low(string):
    up = 0                # Введем две переменные. Это будут счетчики.
    low= 0

    for i in string:          # С помощью цикла for побежимся по всем символам в строке
        if i.isupper():    # Если буква заглавная, то
            up += 1        # Счетчик up увеличивается на единицу
        elif i.islower():  # если буква в нижнем решистре, то
            low += 1       # Счетчик low увеличивается на единицу

    print ('Количество букв в верхнем регистре:',up)    # Выводим на экран подсчитанное количество
    print ('Количество букв в нижнем регистре: ',low)

# Вызываем функцию
up_low('На берегу Амазонки сидел Фернандо')
````

**Задача 2**

Напишите функцию, которая принимает на вход число и проверяет, если оно простое.

Простые числа - это натуральные(целые положительные) числа, которые делятся только на себя или на единицу. 
Пример: 3 - это простое число, т. к. нацело 3 делится только на 3 и на 1.

**Решение:**

````
# Для решения данной задачи воспользуемся алгоритмом нахождения простых чисел - решено Эратосфена
num = int(input("Введите число: "))

def prime(number):
  
  # Проверим что введенное число положительное
  if number < 0:
    print ("Введите число больше нуля")

  # Числа 2 и 3 являются простыми числами, учтем это
  elif (number == 2 or number ==3):
    print (f"Число {number} простое !!!")

  # Проверка что число не простое
  elif (number % 2 == 0 or number % 3 == 0 or number % 3 == 5):
    print (f"Число {number} не простое")

  # В остальных случаях число является простым
  else:
    print (f"Число {number} простое !!!")

# Вызовем функции и в качестве параметра передадим введенное число
prime(num)
````

````
# Можно сделать с помощью другого алгоритма. В результате если число простое вернет True, а если не простое, то False.

def prime2(number):
  i = 2                     # Создадим переменную и присвоем ей значение 2

  # Создаем цикл while. Смысл цикла заключается в том, что пока остаток от деления равен нулю, то цикл продолжается 
  # и будет увеличиваться перменная i на единицу.
  # Если число простое, то цикл будет продолжаться до тех пор пока это число станет равным самому себе. Остаток от деления станет равен 0 
  # и цикл прекратится.

  # Но если число четное, то цикл сразу прекратится, так как условие не выполнется, остаток от деления будет равен нулю. И вернет сразу False
  # Если оно делится на 3 или 5, то цикл прекратится когда i станет 3 или 5 и остаток от деления станет равным нулю и выдаст False.
  
  while number % i != 0:
    i += 1

  # Далее возврат значения. Если введенное нами число равно i, то вернет True, в противном случаи False.
  return number == i

prime2(11)
````

**Задача 3**

Напишите функцию, которая проверяет, если число совершенное.
Совершенным числом является положительное число, такое, что оно является суммой собственных положительных делителей (все, кроме самого числа). 
Пример: первым в числовом ряду совершенным числом является 6: 1+2+3=6
Найдите такие числа в промежутке, который вы будете вводить с клавиатуры (два числа: начало и конец промежутка), например, промежуток от 5 до 150.

**Решение:**

````
# Попросим пользователя ввести два числа. Первый нижняя граница поиска. Второй верхний предел. 

number_down = int(input("С какого числа начать поиск ? "))
number_up = int(input("До какого числа искать ? "))



def perfect_num (num_down, num_up):

  ''' Данная функция ищет совершенные числа в заданном интервале.
      Вход: 
        num_down - с какого числа будет начинатся поиск (нижний предел)
        num_up - до какого числа будет вестись поиск (верхний предел)
      Выход: 
        Печатает числа, если они являются совершенными и находятся в заданном пределе. 
  '''
  
  # Алгоритм поиска будет состоять из двух циклов. В первом будут перебираться все числа в заданном пределе
  for num in range (num_down, num_up):
    sum = 0                                   # У каждого нового числа будет обнулятся сумма (сумма всех его делителей)

  # Во втором цикле каждое число будет проверяться, совершенное ли оно
  # Для этого будем находить сумму всех его делителей.
    for i in range (1, num):
      if num % i == 0:              # Если остаток от деления проверяемого числа будет равняться нулю. То такое число является его делителем.
        sum = sum + i             # Прибавляем к итоговой сумме данное число.
    if num == sum:                  # После того как закончится цикл и найдется сумма всех его делителей, то проверяется равна ли она проверяемому числу.
      print (num)                   # Если равна, то печатаем.


# Вызываем функцию и в качестве параметров передаем то, что ввел пользователь
perfect_num (number_down, number_up)



# Конечно этот алгоритм не совершенный, но вполне рабочий в заданном интервале. И чем больше верхний предел, тем больше времени потребуется для поиска.

# Например в переделах от 1 до 100 -    best - 800 µs per loop
#                      от 1 до 1000 -   best - 33.1 ms per loop
#                      от 1 до 10000 -  best - 3.37 s per loop
#                      от 1 до 100000 - best - 5min 57s per loop
````

**Задача 4**

Напишите функцию, которая проверяет, если фраза является палиндромом. Пример такой фразы:

а роза упала на лапу Азора. 

**Решение:**

````
# Так как представленный пример является палиндромом. И имеет пробелы и разные регистры букв,то
# в начале мы должны убрать пробелы и привести к единому регистру (например нижнему)

# Просим пользователя ввести слово или фразу
str_in = input("Введите строку:")

def pal (string):
  low_str = string.replace(' ','').lower()       # Приводим к единому формату. Без пробелов и в нижнем регистре
  rev_low_str = low_str[::-1]                    # Далее переворачиваем строку и записываем в новую переменную

  if low_str == rev_low_str:                     # Если обе строки одинаковые. Начальная и перевернутая, то 
    print ("Это палиндром!")        
  else:                                          # Иначе
    print ("Не является палиндромом")

pal(str_in)
````

````
# В функции конструкцию if - else можно сократить, использовав return.

str_in = input("Введите строку:")

def pal_2(string):
    string = string.replace(' ','').lower()
    return "Это палиндром!" if string == string[::-1] else "Не является палиндромом"

pal_2(str_in)
````
