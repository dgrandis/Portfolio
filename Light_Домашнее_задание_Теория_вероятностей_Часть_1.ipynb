Задача 1

Магазин получил продукцию с четырех оптовых складов: 5 ящиков с первого склада, 4 со второго, 8 с третьего и 6 с четвёртого. Случайным образом выбран ящик для продажи. Какова вероятность того , что это будет ящик со второго или четвёртого склада.

```
n = 5 + 4 + 8 + 6  # 23 Кол-во исходов 

m_2 = 4      # ящиков со второго склада
m_4 = 6      # ящиков с четвертого склада
```

```
# Исход А 
P = m_2 / n 
print ("Вероятность что ящик со второго склада: ",round(P, 2))
```

```
# Исход В
P = m_4 / n 
print ("Вероятность что ящик с четвертого склада: ",round(P, 2))
```

Задача 4

Найти вероятность того, что при броске двух игральных костей произведение очков

* будет равно 8;
* окажется не менее 23;
* будет нечётным.

```
# Будет равно 8
# Таких исходов будет два 
n = 36
m = [(4,2),(2,4)]
P = len(m) / n 
P
```

```
# Окажется не менее 23
# Вероятные исходы 
n = 36
m = [(6,6),(6,5),(6,4),(5,6),(5,5),(4,6)]
P = len(m) / n 
P
```

```
# Будет нечётным
# На двух кубиках должно быть не четное число
n = 36
m = [(1,1),(1,3),(1,5),(3,1),(3,3),(3,5),(5,1),(5,3),(5,5)]
P = len(m) / n 
P 
```
