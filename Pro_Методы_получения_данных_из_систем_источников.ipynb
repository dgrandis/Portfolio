# Задание **Pro**

Дан RESTFul API https://jsonplaceholder.typicode.com

API предоставляет доступ к 6 ресурсам:

  - /posts 100 posts (Посты пользователей)

  - /comments 500 comments (Комментарии пользователей)

  - /albums 100 albums (Альбомы пользователей)

  - /photos 5000 photos (Фотографии пользователей)

  - /todos 200 todos (Список дел Пользователей)

  - /users 10 users (Список Пользователей)

Используя примеры запросов из прилагаемого ноутбука:

- 1) Получите полный список пользователей

- 2) Получите все посты для пользователя с id=3

- 3) Получите все комментарии пользователя с id=1

- 4) Обновите 3 любых поста для пользователя с id=1

# Подключение библиотек

```python
import json
import requests
```

# 1. Получите полный список пользователей

## Обратимся к словарю **users**, представленному на сайте

image.png

```python
# С помощью модуля requests получаем ссылку на сайт(создаем подключение) для пользователей - users
result = requests.get('https://jsonplaceholder.typicode.com/users')
```

```python
# Получим данные для таблицы users, с помощью библиотеки json и функции loads
users_data = json.loads(result.content)
```

```python
# Отобразим data
print(users_data)
```

Как отображается вывод на сайте: https://jsonplaceholder.typicode.com/users

image.png

Данный вывод не информативный. Поэтому используем библеотеку **PrettyPrinter**

https://docs.python.org/3/library/pprint.html

```python
from pprint import pprint
```

```python
pprint(users_data)
```

Теперь для того чтобы вывести именно пользозователей **user** нужно циклом пройтись по **users_data** и вывести нужную информацию

```python
for i in users_data:
  print(i['name'], i['username'], i['website'], i['email'], i['id'])
```

# 2. Получите все посты для пользователя с id=3

Данные посты находятся в таблице **posts** 

image.png

Пример, как они представлены на сайте

image.png

Теперь нужно только для **третьего useriD** получить список постов.

Согласно документации на сайте, надо вводить вот такой запрос

image.png

```python
result = requests.get('https://jsonplaceholder.typicode.com/posts?userId=3')   # Id=3
```

```python
users_posts = json.loads(result.content)
```

```python
# Выведем содержимое на экран
pprint(users_posts)
```

Так же можно как и из **задания 1** вывести нужную информацию с помощью цикла

```python
for i in users_posts:
  print(i['userId'], i['id'], i['title'])
```

# 3 Получите все комментарии пользователя с id=1

Данные комментарии находятся в таблице **comments** 

image.png

Так же как и предыдущих заданиях с помощью модуля **requests** получаем ссылку на сайт но уже к словарю с комментариями **/comments**

```python
result = requests.get("https://jsonplaceholder.typicode.com/comments/")
```

```python
users_comments = json.loads(result.content)
```

**Посмотрим на результат**


```python
pprint(users_comments)
```

```python
# Далее с помощью цикла так же проходимся по словарю, но уже фильтруя нужную информацию с помощью оператора if

# Так как нам нужны комментарии только пользователей id которого равен 1

for i in users_comments:
  if i['id'] == 1:
    print(i['body'])   # body это комментарии в данном случаи
```

# 4. Обновите 3 любых поста для пользователя с id=1

Мы не можем изменить(обновить) посты на сайте. Но можно вывести данные, а потом их обновить.
Но данные на сайте не изменятся.

```python
result = requests.get('https://jsonplaceholder.typicode.com/posts?userId=1')   # Id=1
```

```python
users_posts = json.loads(result.content)
```

```python
# И выведем на экран результат
pprint(users_posts)
```

Всего **10** комментариев для этого пользователя, а нужно изменить **3**

Будем менять **2**, **4** и **6** комментарии.

```python
for i in [2,4,6]:
  post_change = users_posts[i-1]
  post_change['title']  = f'Комментарий id ползователя {i}: Был изменен на новый'
  data_change = post_change
  requests.put(f'https://jsonplaceholder.typicode.com/posts/{i-1}',data = data_change)

# Посмотрим на измененные комментарии
users_posts 
```

Как видно комментарии под номерами id **2**, **4** и **6** были обновлены.
