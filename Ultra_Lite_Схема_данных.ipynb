# Задание **Ultra Lite**
 

Используя схему данных, состоящую из таблиц:
 - Teachers 
 - Students 
 - Courses
 - StudentCourses
 
Воспроизведите базу данных из занятия. И добавьте в нее таблицу, в которой есть три поля:
  - id (integer)
  - campus (bit)
  - student_id (integer)

Которые говорят о том, проживает ли студент в кампусе или нет. Свяжите эту таблицу с таблицей StudentCourses.

Для выполнения можно пользоваться любой программой по созданию баз данных (https://www.dbdesigner.net , https://www.quickdatabasediagrams.com , https://www.devart.com/ru/dbforge/sql/studio/database-diagram.html и т.д. ). Результат представить в виде картинки любого формата.
