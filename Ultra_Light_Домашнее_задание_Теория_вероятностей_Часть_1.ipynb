Задача 1

Имеется стандартная колода из 52 карт. Рассчитайте, пожалуйста, вероятность
* извлечения короля любой масти,
* извлечения любой карты не пиковой масти.

```
# Извлечения короля любой масти

m = 4
n = 52
P = m / n 
print ("Вероятность извлечения короля любой масти из колоды в 52 карты: ",round(P, 2))
```

Задача 3

В урне находится 15 зеленых кубиков, 11 жёлтых и 7 коричневых. Какова вероятность того, что извлеченный кубик будет
* зеленым;
* жёлтым;
* коричневым.

```
# Всего кубиков в урне: n = 15 + 11 + 7 = 33 (Общее число всех равновозможных, элементарных исходов этого испытания, всего кубиков)

# Так как три действия одинаковые, разница только в значениях, создадим функцию

def prob (m,n,s):
  P = round((m / n), 3)
  print ("Вероятность что извлеченный кубик будет {} цвета: {}".format(s,P))

prob (15,33,'зеленого')
prob (11,33,'желтого')
prob (7,33, 'коричневого')
```

Задача 4

Найдите вероятность того, что при бросании двух игральных костей выпадет
* 8 очков; 
* не более 7; 
* от 9 до 12 включительно.

```
# Всего различных кобинаций 36, n = 6 * 6

n = 36

# 8 очков
m_a = [(2,6), (6,2), (4,4), (5,3), (3,5)]   # Всего комбинаций кубиков для 8 очков

P = round(len(m_a) / n, 2)
print ("Вероятность что выпадет 8 очков составляет: ", P)
```

```
# не более 7 (7 включительно) 
m_b = [(1,1),(1,2),(1,3),(1,4),(1,5),(1,6),(2,1),(2,2),(2,3),(2,4),(2,5),(3,1),(3,2),(3,3),(3,4),(4,1),(4,2),(4,3),(5,1),(5,2),(6,1)]

P = round(len(m_b) / n, 2)
print ("Вероятность что выпадет не более 7 очков составляет: ", P)

# Более 7 (второй вариант решения)
# Узнает вероятность противоположного события:
m_b2 = [(2,6),(3,5),(3,6),(4,4),(4,5),(4,6), (5,3),(5,4),(5,5),(5,6), (6,2),(6,3),(6,4),(6,5),(6,6)]
P = round(len(m_b2) / n, 2)
print ("\nВероятность что выпадет более 7 очков составляет: ", P)

print("Для гипотезы b (не более 7), вероятность: {}".format(round(1 - P, 2)))

```

```
# от 9 до 12 включительно
m_c = [(3,6),(6,3),(4,5),(5,4),(4,6),(6,4),(5,5),(5,6),(6,5),(6,6)]
P = round(len(m_c) / n, 2) 

print ("Вероятность что выпадет от 9 до 12 включительно очков составляет: ", P)
```

Задача 5

В лифт 11-этажного дома на первом этаже зашли 3 человека и поехали. Найдите вероятность того, что
* a) все выйдут на одном этаже;
* b) все выйдут на разных этажах;
* c) двое выйдут на одном этаже.

```
# Сначала определим общее число комбинаций. На 10 различных этажах может выйти каждый пассажир
# n = 10*10*10 = 1000
n = 1000

# Исход А.  Все выйдут на одном этаже.
m_a = 10      # Такой исход всего 1
P = m_a / n 
print ("Вероятность что все выйдут на одном этаже составляет: ", P)
```

```
# Исход B.  Все выйдут на разных этажах.
# Перый пассажир может выйти на любом из 10 этажей, тогда второй только на любом из 9, а третий из 8, поэтому
# Количество элементарных исходов, благоприятствующих событию будет равно 720.
m_b = 10 * 9 * 8 

# Найдем вероятность
P = m_b / n 
print ("Вероятность что выйдут на разных этажах составляет: ", P)
```

```
# Исход C. Двое выйдут на одном этаже

# Так как мы знчем обе другие вероятности. Поэтому можно воспользоваться свойством суммы вероятности проивоположных событий.

P = 1 - (m_a/n + m_b/n )
print ("Вероятность что двое выйдут на одном этаже составляет: ", P)
```
