```
# Подключаем нужные библиотеки
import numpy as np
import matplotlib.pyplot as plt
```

## **Задача 2**

Напишите функцию, которая находит все экстремумы двумерной функции z = sin(x) * sin(y), а также:

a) Возвращает листы с координатами двумерных точек - 4 листа - локальный и глобальный минимум и максимум.

b) Выводит график такой функции.

```
import pylab
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5,5,10)
y = np.linspace(-5,5,10)

#Формируем двумерную сетку
xgrid, ygrid = np.meshgrid(x, y)

#Вычисляем 2d функцию
zgrid = np.sin(xgrid) * np.sin(ygrid)

#Найдем глобальный экстремум

global_mins_x = [] #Глобальный минимум по x
global_mins_y = [] #Глобальный минимум по y
global_maxs_x = [] #Глобальный максимум по x
global_maxs_y = [] #Глобальный максимум по y

#Минимальное и максимальное значение z
min_z = min(min(zgrid.tolist()))
max_z = min(max(zgrid.tolist()))

#Проходим во всем точкам
for i in range(len(x)):
  for j in range(len(y)):
    #Если точка равна минимуму z
    #То это глобальный минимум
    if (zgrid[i, j] == min_z):
      global_mins_x.append(i)
      global_mins_y.append(j)
    
    #Если точка равна максимуму z
    #То это глобальный максимум
    if (zgrid[i, j] == max_z):
      global_maxs_x.append(i)
      global_maxs_y.append(j)

    
#Выводим глобальный минимум
plt.plot([x[i] for i in global_mins_x], [y[i] for i in global_mins_y], 'gs')
  
#Выводим глобальный максимум
plt.plot([x[i] for i in global_maxs_x], [y[i] for i in global_maxs_y], 'rs')

  
#Создаём фигуру и 3D оси
fig = pylab.figure()
axes = Axes3D(fig)

#Отображаем двумерную функцию
axes.plot_surface(xgrid, ygrid, zgrid)
pylab.show()
```
