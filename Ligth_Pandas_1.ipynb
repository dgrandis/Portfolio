### Задача 2
Заполните пустые данные в базе словом "No". Запишите результат в ту же переменную. Выведите ее на экран.

**Решение:**

```
my_t = my_t.fillna("No")                      # С помощью метода fillna заменяем пустые данные словом No и записываем результат в ту же переменную
print (my_t['State'].isna().sum())            # Проверка что не осталось пустых данных в столбце State
my_t                                          # Вывод на экран таблицы с новыми данными
```

### Задание 3

При помощи метода Pandas определите число уникальных записей в столбцах Country, City, Year.

**Решение:**

```
print ("Способ 1\n")
print ("Число уникальных записей в столбце Country:",my_t.Country.nunique(dropna = True)) # Найдем число уникальных элементов в нужных столбцах с помощью метода nunique
print ("Число уникальных записей в столбце City:",my_t.City.nunique(dropna = True))
print ("Число уникальных записей в столбце Year:",my_t.Year.nunique(dropna = True))

print ("\nСпособ 2\n")
print ("Число уникальных записей в столбце Country:",len(my_t['Country'].unique()))   #Найдем число уникальных элементов в нужных столбцах с помощью метода unique и функции len
print ("Число уникальных записей в столбце City:",len(my_t['City'].unique()))
print ("Число уникальных записей в столбце Year:",len(my_t['Year'].unique()))

my_t[['Country', 'City']].describe(include=['O'])                        # Вывод с помощью метода describe для проверки  
```

```
print ("Число элементов в строках которых встречается страна Russia:",my_t[my_t.Country == 'Russia'].size)                           # Определяем с помощью метода size и среза сколько элементов в выборке Russia                   
print ("Сколько всего не пустых записей в строках которых встречается страна Russia:",my_t[my_t.Country == 'Russia'].count().sum())  # В задание не входит. Посмотреть сколько из них заполненных
print ("")                                                                                                                           # Пустая строчка, чтобы отделить от таблицы
my_t[(my_t.Country == 'Russia')]                                                                                                     # Вывод строк где встречается страна Russia
```

### Задача 5
Создайте новый столбец температур, в котором будут записаны те же значения AvgTemperature, но в градусах Цельсия (исходная единица измерения - градусы Фаренгейта).

**Решение:**

```
my_t['Celsius'] = round((my_t['AvgTemperature'] - 32) / 1.8, 1)            # Создаем еще один столбец для значений температуры в градусах Цельсия. 
                                                                           # Переводим по формуле Фарангейты в Цельсия. Округляем до одного знака после запятой и заносим в столбец Celsius
my_t                                                                       # Выводим обновленную таблицу                                        
```
