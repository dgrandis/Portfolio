**Решение:**

### Задание 2

Сложите все значения в таблице по столбцам без использования циклов и условий.

**Решение:**

```
print ("Первый способ:\n")

print ("Сумма значений по столбцам:\n",df.sum())    #Выводим сумму всех столбцов


print ("\nВторой способ:\n")

print ("Сумма 0 столбца:", df[0].sum())              #Выводим сумму каждого столбца в отдельности
print ("Сумма 1 столбца:", df[1].sum())
print ("Сумма 2 столбца:", df[2].sum())
print ("Сумма 3 столбца:", df[3].sum())
print ("Сумма 4 столбца:", df[5].sum())


print ("\nТретий способ:\n")

sum_6 = df[6].sum()                                  # Присваеваем переменной сумму столбца 6 
print ("Сумма 6 столбца:", sum_6)                    # Выводим сумму 6 столбца


```

### Задание 4

При помощи метода Pandas выведите Series, содержащий данные из второго и пятого столбцов.

**Решение:**

```
df.iloc[:,[1,4]]               # С помощью метода iloc и отбора по столбцам выводим второй и пятый столбцы (номера столбцов 1 и 4)
```
