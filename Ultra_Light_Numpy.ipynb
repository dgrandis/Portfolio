### Задача 1

Создайте целочисленный массив Numpy. Для этого выполните следующее:
1. Создайте список из целых значений (не меньше 5 штук).
2. Используя метод numpy.array(), переведите список в массив Numpy.

**Решение:**

```
import numpy as np

# Способ 1

my_list = [23, 88, 126, 49, 287, 92, 7543, 592]
my_array = np.array(my_list)
print (my_array)   

# Способ 2

my_array2 = np.array ([2, 4 ,8, 16, 32, 64, 128, 256])
print (my_array2)   

# Способ 3

my_array3 = np.array([i for i in range(0,30, 3)])
print (my_array3)   
```

###Задача 2

Используя индексы, выведите первый и третий элемент массива.


**Решение:**

```
print ("Первый элемент массива:", my_array[0])
print ("Третий элемент массива:", my_array[2])
```

###Задача 3

**Решение:**

```
print(type (my_array))
print(type (my_array2))
print(type (my_array3)) 
```

###Задача 4

Используя метод sum(), найдите сумму всех элементов массива. 


**Решение:**

```
sum_my_array = my_array.sum()
sum_my_array2 = my_array2.sum()

print("Сумма массива my_array:", sum_my_array)
print("Сумма массива my_array2:", sum_my_array2)
print("Сумма массива my_array3:", my_array3.sum())
```
