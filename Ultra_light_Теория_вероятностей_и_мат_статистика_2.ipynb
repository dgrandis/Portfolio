Задача 1

Цена деления шкалы измерительного прибора равна 0.5. Показания прибора округляются до ближайшего целого деления. Считая, что погрешности округлений распределены равномерно, найти вероятность того, что при очередном измерении она не превзойдёт 0.02.



```
from scipy.stats import norm      # Подключим модуль библиотеки для расчета нормальной непрерывной случайной величины
from scipy.stats import uniform   # Стандартная форма распределения от [0, 1]
from numpy.random import normal   # Библиотека numpy
```

```
b = 0.5

p1 = uniform(0,0.5).cdf(0.02)       # Вероятность что значение в интервале [0, 0.02]
p2 = 1 - uniform(0,0.5).cdf(b-0.02)  # Вероятность что значение в интервале [0.48, 0.5]

P = p1+p2                            # Вероятность что не превзойдет 0.02

print("Искомая вероятность: {}%".format(round(100*P,2))) 
```

Задача 3

Для нормального распределения со средним = 14 и стандартным отклонением = 3 найдите 

* среднее, дисперсию, 
* 50 и 90 перцентили (0.5 и 0.9 квантили). 



```
mu = 14              # Среднее математическое ожидание
std = 3              # Cтандартным отклонение

norm_d = norm(mu,std)  # нормальное распределение

mean, var = norm_d.stats(moments='mv')   # получим среднее и дисперсию 

print ("Среднее: {}". format(mean))
print ("Дисперсия: {}".format(var))

print ("Перцентиль 50: {}". format(norm_d.ppf(0.5)))  # Квантиль 0.5
print ("Перцентиль 90: {}". format(norm_d.ppf(0.9)))  # Квантиль 0.9
```

```
# Видно что значение перцентиля 50% совпадает со средним математическим ожиданием
```

Задача 4

Нарисуйте pdf и cdf нормального распределения со средним = 1.5 и стандартным отклонением = 0.7.


```
from matplotlib import pyplot
mu = 1.5
sigma = 0.7
# Задаем распределение
dist = norm(mu, sigma)

#  pdf
values = [value for value in range(0,4)]
probabilities = [dist.pdf(value) for value in values]
pyplot.plot(values, probabilities)
pyplot.show()
#  cdf
cprobs = [dist.cdf(value) for value in values]
pyplot.plot(values, cprobs)
pyplot.show()
```
