# Задание 1

Дан стэк следующего вида: (1,2,3,’’,5,’g’,’e’,2,5). Распишите шаги и операции,
которые необходимо произвести со **стэком** для извлечения элемента ‘g’.

```python
# Так как этот тип данных является стэком, то применим метод LIFO - Last in first out

# метод Pop - возвращает верхний элемент после удаления из стека

lst = [1,2,3,'',5,'g','e',2,5]

for i in range(len(lst) - lst.index('g')- 1):
  lst.pop()

print(lst[-1])

# Или можно вынести в отдельную переменную
g = lst.pop()
print(g)
```
