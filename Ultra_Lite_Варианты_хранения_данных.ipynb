# Задание 1

ан следующий массив:

a = [1,4,6,1,4,8,10,1,9,2,30].

Написана программа, которая выполняет с массивом следующие операции последовательно:

a.insert(1,1) # формат array.insert(index, element)

a.insert(2,1)

a.insert(6,1)

a.delete(10)

a.delete(4)

Что после проведения этих команд вернет команда a.sort()?

```python
a = [1,4,6,1,4,8,10,1,9,2,30]
```

```python
# Метод insert добавляет элемент в список по номеру индекса

a.insert(1,1)

# Это значит что по номеру индекса 1 добавится целое число 1
```

```python
# Аналогично с другими числами
a.insert(2,1)
a.insert(6,1)
```

```python
# Выведем результат
print(a)
```

```python
# В списках вместо delete используется метод pop()
a.pop(10)
a.pop(4)
```

```python
# Выведем результат
print(a)
```

```python
# Отсортируем полученный список
a.sort()
print(a)
# Сортированный список по возрастанию
```

# Задание 2
Дана очередь следующего вида: (1,2,3,’’,5,’g’,’e’,2,5). Распишите шаги и операции, которые необходимо произвести с очередью для извлечения элемента ‘g’.

```python
lst = [1,2,3,'',5,'g','e',2,5]
```

```python
# Так же воспользуемся методом pop() для удаления элементов из начала списка, так же как Dequeue из лекции.
# Удаление до элемента g будет происходит в цикле

# Для определения индекса 'g' воспользуемся методом index()

for _ in (range(lst.index('g'))):
  lst.pop(0)

# Вывод на экран
print(lst[0])

#  Или можно взять этот элемент из списка
g = lst.pop(0)
print(g)
```
