Импортируем библиотеки matplotlib и seaborn

```
import matplotlib.pyplot as plt       # Импортируем библиотеку mathlotlib
import seaborn as sns                 # Импортируем библиотеку seaborn

%matplotlib inline                    # Это магическая команда, благодаря которой мы получим график рядом с ячейками с кодом
```

### **Задача 1**

На основе данных x и y, представленных ниже, постройте 2 графика plot(x,y) и scatter(x,y) на разных плоскостях.

```
# Данные для построения графика
x = [-7, -5, 7, 9, 10, 10, 9, -7, 0, 5, 7, 4, 0, 6, 8, 4, 0]
y = [0, 2, 2, 5, 5, 1, 0, 0, 2, 6, 6, 2, 1, -3, -3, 1, 1]
```

```
# Построим график с помощью функции plot и в качесте аргументов передаем списки переменных x и y 
plt.plot (x, y)       
plt.show()             # Так как нужно построить два графика на разных плоскостях используем функцию show

# Отобразим график с помощью функции scatter на другой плоскости
plt.scatter (x, y)
plt.show()
```

```
# Так же, если нужно построить два графика на разных плоскостях, то можно воспользоваться функцией subplot

plt.figure(figsize = (6,10))     # Создадим размер общего графика чтобы он соответствовал размерам представленным выше

plt.subplot (211)                # С помощью метода subplot создаем общий макет для будущих графиков.  (2 строки   1 столбец   1 график)
plt.plot (x,y)                   # Построим первый график
 
plt.subplot(212)                 # Аналогично изображаем второй график, но в качестве аргумента передаем 2 (x x 2)
plt.scatter(x, y)                # Построим второй график

plt.show()                       # Фиксируем вывод
```

### **Задача 2**

Добавьте на предыдущий график plot линии сетки, подписи осей х и y, название графика "Самолет". Сделайте размер графика (6, 3).

### **Решение:**

```
plt.figure (figsize =(6,3))                # Делаем график размером 6 на 3
plt.plot (x, y, label ='Самолет')          # Визуализируем график и и можно добавить описание графика - 'Самолет'   
plt.title ('Самолет')                      # С помощью метода title присваеваем название графика 
plt.xlabel ('Ось x')                       # Подписываем ось х
plt.ylabel ('Ось y')                       # Подписываем ось y
plt.grid (True)                            # Отображаем линии сетки
plt.legend()                               # Легенда (lebel)
plt.show()                                 # Фиксируем вывод
```

### **Задача 3**

Постройте, на основе приведенного ниже словаря, диаграмму. Ключи словаря будут подписями на оси х.

```
# Данные для построения диаграммы
data = {'Яблоки': 10, 'Апельсины': 15, 'Лимоны': 5, 'Лайм': 20}
names = list(data.keys())
values = list(data.values())
```

### **Решение:**

```
# Визуализируем диаграмму и в качестве координат стержней (ось х) используем ключи, а в качестве высоты стержней (ось у) их значения из данного словаря

plt.bar (names, values)             
plt.xticks(ticks = names)           # Отображает метки по оси, подписи каждого стержня (подписями будут ключи словаря)

# Так же можно добавить название диаграммы
plt.title ('Кол-во фруктов')    
plt.ylabel ('Количество')           # Можно подписать ось y
plt.xlabel ('Фрукты')               # И ось x общим названием 'Фрукты'

plt.show()                          # Фиксируем вывод
```

### **Задача 4**

Постройте boxplot на основе списка spis, приведенного ниже. Измените цвет графика на зеленый.

```
# Список значений
spis = [2, 4, 1, 5, 6, 13]
```

### **Решение:**

```
# Зададим размер графика
plt.figure(figsize=(6,1))         

sns.boxplot(data=spis, orient = 'h', color ='Green')
plt.show()
# С помощью функции boxplot визуализируем данный список по квартилям
# Цвет графика зеленый
# Ориентация горизонтальная
```

***На графике видно что есть обособленное значение. Это 13 (Отдельная точка)***

```
# Так же можно представить данный график в вертикальной ориентации. 
# И использовать палитру, если бы было несколько входных данных, что бы различать по оттенкам одного цвета
plt.figure(figsize=(1,6))

sns.boxplot(data=spis, palette=sns.color_palette('Greens_r'), orient = 'v')
plt.show()
```

### **Задача 5**

Постройте график lmplot на основе датафрейма, приведенного ниже (строить нужно по данным столбца 'x' и 'y').

```
# Подгружаем датасет "anscombe"
df = sns.load_dataset("anscombe")

# Выводим первые 5 строк датасета
df.head(5)
```

### **Решение:**

```
# Определим парные взаимосвязи на графике с помощью функции lmplot()

sns.lmplot(data=df, x = 'x', y = 'y')
plt.show()

# Ниже есть еще решения данной задачи
```

```
# Так же входные данные х и у можно задать в другом виде

sns.lmplot(data=df, x=df.columns[1], y=df.columns[2])
plt.show()
# Видим точно такой же график
```

```
# Можно раскрасить точки на графике в разные цвета, которые будут определяться количеством уникальных значений в колонке dataset

sns.lmplot(data=df, x=df.columns[1], y=df.columns[2], hue ='dataset')
plt.show()
```
