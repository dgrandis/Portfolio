Создайте единичную матрицу размером 5*5.

###Задача 2

Создайте массив из случайных значений из стандартного нормального распределения размера 6 на 6. 

Разверните массив таким образом, чтобы его размер стал 36 на 1.
Найдите среднее значение элементов массива.

**Разрешается использовать только методы Numpy. (Циклы и условия запрещены)**.

**Решение:**

```
normal_array = np.random.normal (1, 10, (6,6))                # Создание массиива из случайных значений из стандартного нормального распределения размера 6 на 6.
normal_array_new = normal_array.reshape (36,1)                # Изменяем размерность массива 
print ("Размерность массива:", normal_array_new.shape)        # Проверка размерности массива
mean = normal_array_new.mean()                                # Находим среднее значение массива с помощью метода mean и присваеваем переменной
print ("Среднее значение данного маасива:", mean)             # Выводим на экран значение переменной в короторой хранится среднее значение массива 

print ("Другим способом вывести:", normal_array_new.mean())   # Второй способ, без переменной, сразу на экран
```

###Задача 3

Создайте массив из 20 любых случайных значений, используя метод Numpy.
При помощи срезов и операций с массивами найдите значения производных для созданного массива

**Решение:**

```
my_array_3 = np.random.randint(0, 10, (20))                     # Создадим массив размером 20 из случайных чисел

print ("Созданный массив:", my_array_3)                         # Вывести массив на экран
print ("Размер массива:", my_array_3.size)                      # Размерность созданного массива


slice_array = my_array_3[1:]                                    # Создаем массив а со срезом (начиная со втрого элемента)
print ("\nМассив со срезом, начиная со второго элемента:",slice_array)    
print ("Размер массива:", slice_array.size)                        

der = slice_array - my_array_3[:-1]                               # Вычетаем из массива (со срезом) изначально созданный массив (кроме последнего элемента) 
print ("\nЗначения производных:",der)
print ("Размер массива:", der.size) 


#Второй способ

der_2 = np.diff(my_array_3)                                           # Вычисляем производную данного массива. Результат присваеваем переменной der_2 
print ("\nПроизводная массива вторым способом с помощью diff:", der_2)                 # Вывод на экран нового массива (размерность -1)
print ("Размер массива:", der_2.size)                                 # Размерность массива


```

###Задача 4

Дано выражение

Y = A * X * C

Создайте три случайных массива A, C, X. Размерности должны позволять выполнить матричное умножение согласно формуле. На выходе должен получиться массив Y с размерностью (5, 18). (Остальные размерности возьмите на свое усмотрение).

**Решение:**

```
A = np.random.randint (0, 5, (5,3))              # Создаем массив размерностью 5 на 3 из случайных чисел
X = np.random.randint (0, 5, (3,2))              # Создаем массив размерностью 3 на 2 из случайных чисел
C = np.random.randint (0, 5, (2,18))             # Создаем массив размерностью 2 на 18 из случайных чисел

mult = A.dot(X)                                  # Выполняем матричное умножения массивов А на Х. Переменной mult присваеваем результат. 
Y = mult.dot(C)                                  # Выполняем матричное умножения массива mult и С. Переменной Y присваеваем результат. 

print (Y)                                        # Вывод на экран массив Y 
print (Y.shape)                                  # Проверка размерности массива Y
```
