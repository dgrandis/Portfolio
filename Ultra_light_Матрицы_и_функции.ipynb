#Задача 3

Возьмите переменную с матрицей из предыдущих заданий.
С помощью соответствующих методов:

* Транспонируйте матрицу
* Найдите обратную матрицу
* Матрично умножьте саму на себя

**Решение:**

```
# 1

# Выведем первоначальную матрицу
print ("Первоначальная матрица: \n",array)
print ("")

# Транспонируем матрицу
array_trans = array.transpose()
print ("Методом transpose:\n ", array_trans)

# Так же можно с помощью более простого написания
array_t = array.T
print ("\nМетодом .T:\n ", array_t)


# 2

# Найдем обратную матрицу с помощью специального метода np.linalg.inv()
array_inverse = np.linalg.inv(array)
print ("\nОбратная матрица:\n ", array_inverse)

# 3

# Матрично умножим матрицу саму на себя
# Сразу вывдем на экран

print ("\nМатричное умножение матрицы array самой на себя:\n\n С помощью метода .dot:\n",array.dot(array))

print("\nС помощью оператора @:\n", array @ array)
```
