# Подключение библиотек

```python
import requests
import json
```

# Ключ API

```python
api_key = 'Сюда вставить ключ API'
```

# Список адресов

```python
addres = ['ул. Берзарина, 30, Москва, 123060', 'ш. Энтузиастов, 56 строение 35, Москва, 111123', 'Кедрова ул., 6 корпус 3, Москва, 117292']
```

# Получение информации

Документация 

https://developers.google.com/maps/documentation/places/web-service/search-find-place

image.png

image.png

```python
res = []  #  Создаем пустой список где будет хранится словарь json

# Проходимся циклом по списку с адресами, чтобы их подставлять в get запрос

for i in addres:
  r = requests.get(f'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={i}&inputtype=textquery&fields=photos,formatted_address,name,rating,opening_hours,geometry&key={api_key}')
  res.append(json.loads(r.content))
```
