```
# Если считать что ладьи две, первая ладья может стоять на любой клетке
# Тогда для второй ладьи может быть только 14 клеток,чтобы она стояла на одной линии с первой
# Так как первая ладья уже стоит, тогда свободных клеток будет только 63.
14 / 63     
```

```
# 1        - Вероятность первой ладьи
# 49/63    - Для второй ладьи того же цвета
# 2/62   - Есть только две клетки для ладьи другого цвета (чтобы начал образововаться прямоугольник)
# 1/61   - Вероятность для четвертой ладьи 
1 * 49/63 * 2/62 * 1/61   # Вероятность того что 4 ладьи будут бить друг друга (прямоугольник)
```

Задача 3

Подбрасывается 20 монет. Найти вероятность того, что

* на всех монетах выпадет решка;

* орёл выпадет КАК МИНИМУМ на половине монетах. (Для решения этой части используйте формулу Бернулли).

```
# Возьмем функции из комбинаторики
def P(n): #(факториал)
  f =1
  for x in range(2,n+1):
    f*=x
  return f

# Сочетание 
def C_n_m(n,m):
  return P(n)/ (P(n-m)* P(m))
```

```
# На всех монетах выпадет решка
# 1/2 - Вероятность что выпадет решка при одном броске
# При 20 случаях нужно 1/2 * 1/2 * 1/2......1/2 (20 раз) Или возвести в степень 20
1 / 2 ** 20 
```

```
# Создадим функцию по формуле Бернулли
def Bernoulli(m,n,p):
  '''
      Вход:
      m - Количество исходов благоприятствующих событию
      n - Полное количество исходов
      p - Вероятность благоприятных исходов
      Р() - функция нахождения факториала
  '''
  return P(n)/ (P(n-m)* P(m)) * p**m * (1-p)**(n-m)   # Формула Бернулли
```

```
# Вызовем функцию 
Bernoulli(10,20,0.5)            # Формула Бернули что ровно в 10 случаях будет решка
```

```
C_n_m(20,10) * 1 /1048576   # Вероятность ровно в 10 случаях (проверка вручную по формуле Бернули)
```

```
# Пробегаемся по циклу для того что бы рассмотреть все случаи когда решка и в 10 случаях и в 11 и в 12 и в 13 и так далее до 20
p = 0
for i in range (10,21):
  p += Bernoulli(i,20,0.5)

p
```

Задача 4

Какова вероятность того, что в 5 сданных картах будет один туз и один король? (Колода 36 карт).

```
def P(n): #(факторил)
  f =1
  for x in range(2,n+1):
    f*=x
  return f

# Сочетание без повторений
def C_n_m(n,m):
  return P(n)/ (P(n-m)* P(m))
```

```

R5 = C_n_m(36, 5)  # кол-во возможных раздач 5 карт из колоды 36 карт
print(R5)

R3 = C_n_m(28, 3)   # сколько способов раздать 3 карты не содержащие туза или короля из колоды
print(R3)

R1 = C_n_m(4, 1)    # сколько способов раздать 1 короля или туза
print(R1)

R = R1 * R1 * R3    # способов раздать из колоды 1 туза, 1 короля и 3 других любых карты
print(R)

p = R / R5     # вероятность раздать 1 туза, 1 короля и 3 других любых карты
print(p)
```
