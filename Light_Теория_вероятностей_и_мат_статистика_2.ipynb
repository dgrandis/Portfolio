Задача 1

Случайная величина  ошибки взвешивания распределена по нормальному закону с нулевым математическим ожиданием и стандартным отклонением 4 грамма. Найти вероятность того, что очередное взвешивание будет проведено с ошибкой, не превышающей по модулю 6 граммов.

```
from scipy.stats import norm      # Подключим модуль библиотеки для расчета нормальной непрерывной случайной величины
from scipy.stats import uniform   # Стандартная форма распределения от [0, 1]
from numpy.random import normal   # Библиотека numpy
from matplotlib import pyplot     # Библиотека для построения графиков matplotlib
```

```
mu = 0               # Среднее математическое ожидание
std = 4              # Cтандартным отклонение

norm_1 = norm(mu,std)  # нормальное распределение

cdf1 = norm_1.cdf(6)
cdf2 = norm_1.cdf(-6)

print (cdf1)    # вероятность попадания в интервал от (-inf до 6)
print (cdf2)    # вероятность попадания в интервал от (-inf до -6)

print(cdf1-cdf2)
```

Задача 2

Автобусы некоторого маршрута идут строго по расписанию и интервалом 5 минут. Составьте функцию плотности случайной величины  X – времени ожидании очередного автобуса пассажиром, который наугад подошёл к остановке. Найдите вероятность того, что он будет ждать автобус не более 2 минут. 

```
stand_2 = uniform(0,5)    # плотность 1/5 на интервале [0, 5]
p1 = stand_2.cdf(2)       # Вероятность что значение в интервале [0, 2]

print(p1) 
```

Задача 3

Рост мальчиков возрастной группы 15 лет есть нормально распределённая
случайная величина X с математическим ожиданием роста 161 см и среднеквадратичным отклонением 4 см. Какую долю костюмов для мальчиков, имеющих рост от 152 до 158 см, нужно предусмотреть в объёме производства для данной возрастной группы? 


```
mu = 161              # Среднее математическое ожидание
std = 4               # Cтандартным отклонение

norm_3 = norm(mu,std)  # нормальное распределение

cdf1 = norm_3.cdf(158)  # вероятность попадания в интервал от (-inf до 158)
cdf2 = norm_3.cdf(152)  # вероятность попадания в интервал от (-inf до -152)

print (cdf1)    
print (cdf2)    

print(cdf1-cdf2)
```

Задача 4

На машиностроительном заводе в процессе производства за неделю обнаруживается в среднем 100 бракованных деталей со среднеквадричным отклонением в 12 деталей. 
Рассчитайте следующее:
1)
вероятность того, что за неделю будет 20-30 бракованных деталей,
2)
вероятность того, что за день будет хотя бы одна забракованная деталь.

```
mu = 100              # Среднее математическое ожидание
std = 12              # Cтандартным отклонение

norm_4 = norm(mu,std)  # нормальное распределение

cdf1 = norm_4.cdf(30)  # вероятность попадания в интервал от (-inf до 30)
cdf2 = norm_4.cdf(20)  # вероятность попадания в интервал от (-inf до 20)
cdf3 = norm_4.cdf(7)   # вероятность попадания в интервал от (-inf до 7)

print(cdf1)    
print(cdf2)    

print(cdf1-cdf2)      # вероятность того, что бракованных деталей будет от 20 до 30
print(1-cdf3)         # вероятность того, что бракованных деталей за неделю будет как минимум 1 в день
```

* Для всех предыдущих задач вычислите среднее, дисперсию и 0.25 и 0.9 перцентили.

* Для всех предыдущих задач постройте графики cdf и pdf.

```
mean_1, var_1 = norm_1.stats(moments='mv')
perc_1_25 = norm_1.ppf(0.25)
perc_1_90 = norm_1.ppf(0.90)
print ("Среднее: {}".format(mean_1))
print ("Дисперсия: {}".format(var_1))
print ("Перцентиль 0.25: {}".format(perc_1_25))
print ("Перцентиль 0.90: {}".format(perc_1_90))

#  pdf
values_1 = [value_1 for value_1 in range(-10, 10)]
probabilities_1 = [norm_1.pdf(value_1) for value_1 in values_1]
pyplot.plot(values_1, probabilities_1)
pyplot.show()

#  cdf
cprobs_1 = [norm_1.cdf(value_1) for value_1 in values_1]
pyplot.plot(values_1, cprobs_1)
pyplot.show()
```
