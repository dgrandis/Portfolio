# Задание **Ultra Lite**

Придумайте (или возьмите из предыдущих ДЗ) схему базы данных.

Придумайте под нее бизнес модель, в которой существуют по крайней мере 2 отдела (маркетинг и продажи). 

Предложите по одной витрине данных для каждого отдела.

image.png

Данная схема упрощенно представляет базу данных организации, имеющей 5 отделов: маркетинг, логистика, продажи, закупок, управление запасами. 

Отдел продаж центральный и для эффективности важен контроль и поддержание необходимого кол-ва остатков и он взаимодействует с отделом маркетинга и с отделом закупа, а также управление запасами и логистика. А также анализ конкурентноспособности товара.

Для маркетинга важна цена у поставщика, цена товара у конкурентов а также анализ рыночной потребности в товаре он взаимно зависим с отделом закупа.

Управление запасами получает информацию от отдела продаж и передает информацию в отдел закупа
