# Задание:

Попытаться доказать утверждение центральной предельной теоремы о том, что сумма N независимых и одинаково распределенных случайных величин, отобранных практически из любого распределения, имеет распределение, близкое к нормальному. Для доказательства взять семплирования из семейства экспоненциальных распределений.

Математически доказывать не нужно, достаточно проиллюстрировать графически верность утверждения.

Записать вывод о том, как размер выборки влияет на доказательство.

# Подключение бибилиотек

```python
import numpy as np
from matplotlib import pyplot as plt
```

Возьмем произвольное экспанециальное распределение

```python
plt.hist(np.random.exponential(10, 5000), bins=50)
plt.show()
```

Берем произвольное число элементов из распределения и суммируем его несколько раз. 
Так проводим несколько экспериментов и строим гистограмму. По одной оси значение суммы, а по другой сколько эта сумма встречается.

В цикле выведем **10** графиков. И для каждого графика будем выводить результат суммирования нашего распределения несколько раз.
Меняться будет количество экспериментов суммирования. От этого будут меняться гистограммы.

 - Будем брать **1000** чисел и их суммировать.
 - Экспонициальное распределение **100** чисел.

Теперь можно посмотреть как распределение будет меняться при увеличении числа экспериментов для суммирования.

```python
Numbers = 1000
for i in range(1,11):
  num = [np.random.choice(np.random.exponential(5, 100), Numbers).sum() for _ in range(2**i)]
  plt.hist(num, bins = 50)
  plt.show()
```

**Как видно чем больше экспериментов, тем ближе подходит распределение суммы на нормальное распределение на нормальное.** 

Если зафиксировать количество экспериментов на сумму: **const_exp = 10** 
И будем уже менять размер выборки

```python
const_exp = 10   # Количество экспериментов

Numbers = 1000
for i in range(1,11):
  num = [np.random.choice(np.random.exponential(5, 2**i), Numbers).sum() for _ in range(const_exp)]
  plt.hist(num, bins = 50)
  plt.show()
```

# Вывод:

1. Размер выборки **не влияет** на доказательство центральной предельной теоремы.
2. Чем **больше** экспериментов, тем **ближе** подходит распределение суммы на нормальное распределение на нормальное.
3. Влияет число экспериментов, а не размер выборки.
