### Задача 2

Напишите код, который считает среднее значение всех элементов массива (без использования встроенных функций np.mean и т. д.). Для решения задачи воспользуйтесь циклом.
* С помощью команды %%timeit отобразите время выполнения написанного кода.
* Преобразуйте массив my_array в список (с помощью метода list()), сохраните его в переменной my_list.
* В новой ячейке посчитайте среднее значение всех элементов списка (воспользуйтесь тем же кодом, что и для массива).
* Командой %%timeit отобразите время выполнения и этой ячейки.

**Решение:**

```
%%timeit -n 5 -r 10                                  

mean_my_array = 0                                    # Переменная которая будет хранить среднее значение всех элементов массива
sum_my_array = 0                                     # Переменная которая будет хранить сумму всех элементов массива

for i in my_array:
  sum_my_array += i                                  # Ищем сумму всех элементов массива my_array
  mean_my_array = sum_my_array / my_array.size       # Находим среднее значение разделив общую сумму на кол-во элементов в массиве

print ("Среднее значение всех элементов массива равна:", mean_my_array)

my_list = list(my_array)                             #Преобразование массива в список 


```

```
%%timeit -n 5 -r 10

sum_my_list = 0
mean_my_list = 0

for i in my_list:
  sum_my_list += i                                                     # Вычисляем сумму всех элементов в списке
  mean_my_list = sum_my_list / len(my_list)                            # Так же находим среднее значение всех элементов в списке

print ("Среднее значение всех элементов массива равна:", mean_my_list)
```

### Задача 4 

* Используя библиотеку numpy, создайте массив 3x3, значения которого находятся в диапазоне от 11 до 40 (раздел "Семплирование из распределений"). 

* Не используя цикл for,  выведите на экран список элементов, которые меньше 20 (раздел "Индексация").

* Просуммируйте все элементы, не используя цикл for (раздел "Встроенные методы").


**Решение:**

```
my_3x3_array = np.random.randint (11,40, (3,3))                                             # Создаем двумерный массив (матрицу) 3х3 с произволными значениями в диапазоне от 11 до 40
print (my_3x3_array)                                                                        # Просто вывести массив для наглядности

print ("Элементы значения которых меньше 20:", my_3x3_array[my_3x3_array < 20])           # Вывести все элементы списком значения которых меньше 20

sum_my_3x3_array = my_3x3_array.sum()                                                        # Находим сумму всех элементов данного массива
print ("Сумма всех элементов массива равна:", sum_my_3x3_array)

#Или можно сразу так
print ("Сумма всех элементов массива равна:", my_3x3_array.sum())
```

### Задача 5

* Создайте массив, который будет содержать списки с именем студента(str), его возрастом(int) и средней оценкой(float) (обратите внимание на пример с занятия со списком автомобильных регионов).
* Отсортируйте такой массив (обратите внимание на параметр order метода sort()).
* Замените значение, отвечающее за возраст, у всех студентов на одно и то же, например, 10. Отсортируйте такой массив. 

**Решение:**

```
my_type = [('name', '<U32'), ('age', np.int32), ('rating', np.float64)]

students_list = [('Феодул', 22, 4.85), ('Агафангел', 24, 4.37), ('Гермоген', 21, 3.97), ('Епимах', 27, 4.18), ('Каллистрат', 18, 3.55), ('Павсикакий', 23, 4.05), ('Ферапонт', 30, 4.99)]

students_array = np.array (students_list, dtype = my_type)     #Создаем numpy-массив из указанных значений и задаем тип данных для каждого столбца

students_array.sort(order = 'age')
print (students_array,'\n')                                          #Печать отсортированного массива по возрасту

#for i in students_array:                                     #Заменим возраст всех студентов на значение 10
#  i[1] = 10

students_array['age'] = 25

print ("Сортированный список после изменения значения возраста на 10\n")

students_array.sort(order = 'age')                            #Снова отсортируем массив по возрасту
print (students_array)                                        #Печать вновь отсортированного массива
```
