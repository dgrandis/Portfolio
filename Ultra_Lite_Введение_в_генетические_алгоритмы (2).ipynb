# Подключение библиотек

```python
import random as random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
%matplotlib inline 
```

# Функции

```python
# Функция получения выжившей популяции
def getSurvPopul(popul, val, nsurv, reverse=False):
  newpopul = []
  sval = sorted(val, reverse=reverse)
  for i in range(nsurv):
    index = val.index(sval[i])
    #print(index)
    newpopul.append(popul[index])
  return newpopul, sval
```

```python
# Функция получения родителей
def getParents(curr_popul, nsurv):
  indexp1 = random.randint(0, nsurv - 1)
  indexp2 = random.randint(0, nsurv - 1)
  botp1 = curr_popul[indexp1]
  botp2 = curr_popul[indexp2]
  return botp1, botp2
```

```python
# Функция смешивания (кроссинговера) двух родителей
def crossPointFrom2Parents(botp1, botp2, j):
  pindex = random.random()
  if pindex < 0.5:
    x = botp1[j]
  else:
    x = botp2[j]
  return x
```

# Основные данные

```python
n = 100           # Размер популяции
nsurv = 20        # Количество выживших
nnew = n - nsurv  # Количество новых - (80)
l = 12            # Длина бота
epohs = 1000      # Количество эпох
mut = 0.25        # Коэфициент мутаций

# Создадим таблицу
df_result = pd.DataFrame(columns = ['Размер популяции', 'Коэффициент мутации', 'Количество выживших', 'Значение функции f=']) 
```

# Реализация

```python
popul = []                                # Двумерный массив популяции
val = []                                  # Одномерный массив значений этих ботов

# Создаем ботов
for i in range(n):
  popul.append([])
  for j in range(l):
    popul[i].append(random.random())

# Проходим по всем эпохам
for ep in range(epohs): 
  val = []                               # Пустой список для значений ботов

  for i in range(n):                     # Проходим по всей длине популяции
    bot = popul[i]

    x = bot[0] + 5*bot[1] + 10*bot[2]  +  25*bot[3]
    y = bot[4] + 5*bot[5] + 10*bot[6]  +  25*bot[7]
    z = bot[8] + 5*bot[9] + 10*bot[10] +  25*bot[11]

    f = x**2/y + z*(12-z) + 5*y/z       # Функция

    val.append(abs(f))
    #print('val: ', val)

  newpopul, sval = getSurvPopul(popul, val, nsurv)  # Получаем новую популяцию и сортированный список значнией
  print(ep, " ", [round(s,8) for s in sval[0:5]])   # 5 лучших ботов

  for i in range(nnew):
    botp1, botp2 = getParents(newpopul, nsurv)      # Получаем двух случайных родителей
    newbot = []

    for j in range(l):                              # Проходим по всей длине бота
      x = crossPointFrom2Parents(botp1, botp2, j)
      x += mut * (2 * random.random() - 1.0)
      newbot.append(x)
    newpopul.append(newbot)

  popul = newpopul
```

# Запись результатов в таблицу Pandas

```python
# Зададим списки 
population_size    = [50, 100, 150]       # Размер популяции
quantity_survivors = [15, 20, 25]         # Количество выживших
mutation_rate      = [0.2, 0.25, 0.3]     # Коэффициент мутации

l = 12            # Длина бота
epohs = 1000      # Количество эпох
count = 1         # Счетчик экспериментов
```

```python
# Создадим таблицу
df_result = pd.DataFrame(columns = ['Размер популяции', 'Количество выживших', 'Коэффициент мутации', 'Значение функции f='])
```

```python
for p_s in population_size:         # [50, 100, 150]
  for q_s in quantity_survivors:    # [15, 20, 25]
    nnew = p_s - q_s                # Количество новых (популяция минус выжившие)
    for m_s in mutation_rate:       # [0.2, 0.25. 0.3] 

      popul = []                                
      val = []                                  

      # Создаем ботов
      for i in range(p_s):
        popul.append([])
        for j in range(l):
          popul[i].append(random.random())

      # Проходим по всем эпохам
      for ep in range(epohs): 
        val = []                               

        for i in range(p_s):                     
          bot = popul[i]

          x = bot[0] + 5*bot[1] + 10*bot[2]  +  25*bot[3]
          y = bot[4] + 5*bot[5] + 10*bot[6]  +  25*bot[7]
          z = bot[8] + 5*bot[9] + 10*bot[10] +  25*bot[11]

          f = x**2/y + z*(12-z) + 5*y/z       

          val.append(abs(f))
          #print('val: ', val)

        newpopul, sval = getSurvPopul(popul, val, q_s)  
        #print(ep, " ", [round(s,8) for s in sval[0:5]])   

        for i in range(nnew):
          botp1, botp2 = getParents(newpopul, q_s)      
          newbot = []

          for j in range(l):                              
            x = crossPointFrom2Parents(botp1, botp2, j)
            x += m_s * (2 * random.random() - 1.0)
            newbot.append(x)
          newpopul.append(newbot)

        popul = newpopul

      # Запись результата
      bot = popul[0]

      x = bot[0] + 5*bot[1] + 10*bot[2]  +  25*bot[3]
      y = bot[4] + 5*bot[5] + 10*bot[6]  +  25*bot[7]
      z = bot[8] + 5*bot[9] + 10*bot[10] +  25*bot[11]

      f = x**2/y + z*(12-z) + 5*y/z

      # Добавление в таблицу
      df_result.loc[f'{count}'] = [p_s, q_s, m_s, abs(f)]
      count += 1      # Увеличение счетчика для записи в таблицу
```
