```
def P(n): #(факторил)
  f =1
  for x in range(2,n+1):
    f*=x
  return f

# Сочетание 
def C_n_m(n,m):
  return P(n)/ (P(n-m)* P(m))

# Размещение 
def A_(n,m):
  return C_n_m(n,m) * P(m)

# Перестановка с повторением
def P_rep(n,m):
  add = 1
  for x in m:
    add*=P(x)
  return P(n)/ add

# Сочетание с повторением
def C_n_m_rep(n,m):
  return P(n+m-1)/ (P(n-1)* P(m))

# Размещение с повторением
def A_rep(n,m):
  return n**m


```

###Задача 1
 
  Записать множество Е, если $E = A\cap B $, причем А={2, 4, 6, 8, 10, 12}, B={3, 6, 9, 12}.

```
# Данная задача на пересечение(И) множеств А и В. 
A = {2, 4, 6, 8, 10, 12}
B = {3, 6, 9, 12}
E = {e for e in A if e in B}   # В переменную Е будут добавлять только те элементы, которые есть в обоих множествах А и В. 
E
```

###Задача 2
 
  Записать множество Е, если $E = A\cup B $, причем А={1, 6, 2, 7, 9, 3}, B={1, 6, 2, 7, 9, 3}.

```
# Задача на объединение(ИЛИ) множеств. 
E = A.copy()
E.update(B)
E
# В переменной Е будут множество из А и В.
```

###Задача 3
 
  Записать множество Е, если $E = A\B $, причем А={2, 4, 6, 8, 10, 12}, B={1, 3, 6, 9, 12}.

```
# Задача на разность(И). Каждый элемент которого принадлежит множеству А и не принадлежит множеству B.
A = {2, 4, 6, 8 ,10, 12}
B = {1, 3, 6 ,9 ,12}
E = {e for e in A if e not in B}
E 
```
