# Импорт необходимых библиотек

```python
import requests
```

## Ноутбук из урока

**Возьмем пример из урока и будем добавлять и изменять необходимые параметры**

```python
endpoint = "https://en.wikipedia.org/w/api.php"
action = "?action=query&format=json&generator=search&gsrsearch="
criterion = 'Tramp'

res = requests.get(endpoint+action+criterion)
res.json()
```

## 1. Добавьте в ответ текущий timestamp

```python
# Надо изменить строку action где мы будем получать итогую запись
# Параметры записываются через значок амперсанда - &
# Поэтому надо добавить новую запись  - timestamp

# timestamp - дата когда был сделал запрос
```

image.png

```python
# Добавим curtimestamp это в строку через &

action = "?action=query&curtimestamp&format=json&generator=search&gsrsearch="
```

## 2. Измените формат ответа на xml

**Формат был json станет xml**

```python
#Изменим json на xml

action = "?action=query&curtimestamp&format=xml&generator=search&gsrsearch="
```

## 3. Выведите количество слов в документе (wordcount)

**Первое надо поменять парметр на list=search**

image.png

```python
action = "?action=query&curtimestamp&format=xml&list=search&gsrsearch="
```

И добавим параметр **srprop=wordcount** так же через &

image.png

```python
action = "?action=query&curtimestamp&format=xml&list=search&srprop=wordcount&srsearch="
```

# Результат

```python
endpoint = "https://en.wikipedia.org/w/api.php"

action = "?action=query&curtimestamp&format=xml&list=search&srprop=wordcount&srsearch="

criterion = 'Trump'

res = requests.get(endpoint+action+criterion)
print(res.text)
```

- Как видно в результате вывод в виде **xml**
- Добавилось время запроса
- И количество слов на каждой странице
