```python
import pika
```

```python
# подключились к брокеру сообщений, находящемуся на локальном хосте
connection = pika.BlockingConnection(pika.ConnectionParameters('localhost'))
channel = connection.channel()
```

```python
# создаём очередь, в которую будет отправлено сообщение, назовем очередь «hello»
channel.queue_declare(queue='hello')
```

```python
# отправляем сообщение 'Hello World!' в очередь 'hello'
channel.basic_publish(exchange='',
                      routing_key='hello',
                      body='Hello World!')
print (" [x] Sent 'Hello World!'")
```

```python
connection.close() # закрываем соединение
```
