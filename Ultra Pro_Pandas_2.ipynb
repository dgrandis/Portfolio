### Задача 1
Подгрузите данные 'data.csv' (возможно, придется указать кодировку encoding сp1250) в переменную data.

Создайте новый столбец TotalCost со значениями Quantity * UnitPrice. Сделайте resampling данных в еженедельные со значениями суммы по столбцам TotalCost.


**Решение:**

```
import pandas as pd                        # Подключаем библиотеку pandas 
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
data = pd.read_csv('/content/drive/MyDrive/NU/04 Pandas/data.csv', encoding = 'cp1250')

# В качестве индекса используем колонку InvoiceDate преобразованную в DateTime index
data.index = pd.to_datetime(data.InvoiceDate)     

# Удаляем колонкуInvoiceDate
data.drop('InvoiceDate', axis =1, inplace = True)
```

```
#Добавим новую колонку TotalCost со значениями Quantity * UnitPrice (Количество * Цена)
data['TotalCost'] = data['Quantity'] * data['UnitPrice']

# Посмотрим полученный результат
data
```

```
#Resampling суммы (sum) TotalCost
# Начало ('W') недели (если не указать то по умолчанию будет воскресенье SUN) И сумма за неделю будет считаться с понедельника по воскресенье включительно
# Но если поставить ('W - MON') понедельник, то итоговая сумма будет считаться со вторника по понедельника следующей недели включительно.

data.TotalCost.resample ('W').sum()
```

### Задача 2

Посчитайте скользящее недельное среднее суммы ежедневных заказов. Для этого
* сделайте resampling в ежедневные данные,
* посчитайте скользящее среднее за 7 дней по столбцу TotalCost.

*Для решения второй части данной задачи потребуется самостоятельно найти способ определения скользящего среднего.*

**Решение:**

```
# Сделаем resampling по дням
day_total = data.TotalCost.resample('D').sum()

# Посмотрим результат
print ("Сумма заказов по дням(head):\n",day_total.head())

#Сделаем resampling по неделям и разделим на 7. Получится среднее значение по всем неделям
print ("\nСреднее скользящее по неделям(head):\n")

# Берем переменную day_total
(day_total.resample('W').sum() / 7).head()
```

```
# Или можно сделать проще,
# как можно подсчитать среднюю сумму ежедневных заказов за каждую неделю. 
# Для этого сумму за неделю разделим на 7 

# Например, если брать неделю с 2010-12-06 по 2010-12-12, 
# то по дням будет (53860.18 + 45059.04 + 44189.84 + 52532.13 + 57404.91 + 0 + 17240) - просто для проверки
# И подсчитаем среднее значение разделив эту сумму на 7 дней в неделе, получим за неделю 2010-12-12 : 38612.43 (вторая строка)

week_mean = data.TotalCost.resample('W').sum() / 7

# Выведем результат за все время по неделям
week_mean.head()

# Результат такой же
```

```
# Так же можно подсчитать среднее значение каждого заказа по дням и неделям
# Найдем среднее значение по дням
day_mean = data.TotalCost.resample('D').mean()

# В итоге видим. Например,что в среднем каждый заказ 6 декабря был на 13.88 , а 7 декабря на 15.20 и так далее
print("Пример одной недели:\nСредняя цена за заказ по дням:\n\n",day_mean['2010-12-06' : '2010-12-12'])

# Теперь подсчитам среднее значение по неделям исходя из средних значений за день.

print ("\nСредняя цена заказа в каждой конкретной недели:\n")
data.TotalCost.resample('W').mean()

# Например видим, что среднея цена одного заказа за неделю 2010-12-12 была: 16.29 (вторая строчка)
```

### Задача 3. GroupBy

Посчитайте общую стоимость заказов на ежемесячной основе в разбивке по странам.


**Решение:**

```
# Сгруппируем по странам. И в каждой отдельной стране посчитаем общую сумму заказов за каждый месяц.
country_month = data.groupby('Country').TotalCost.resample('M').sum()

# Выведем сохраненный результат
country_month
```

```
# Так же можно посчитать общую сумму за год по каждой отдельной стране  (К задаче не относится)
data.groupby('Country').TotalCost.resample('Y').sum()
```

### Задача 4. JOIN

* Подгрузите данные, содержащие информацию о номерах карт лояльности пользователей (card_data.csv).
* Назовите ее cards.
* Сделайте Left Join исходной таблицы (которая получилась после задачи с datetime индексом) и таблицы cards.
* Если будет переполнение памяти, попробуйте взять не все значения таблицы, а, например, первые 100 с помощью соответствующего метода iloc.

**Решение:**

```
cards = pd.read_csv('/content/drive/MyDrive/NU/04 Pandas/card_data.csv')

# Выведем dataFrame, чтобы посмотреть какие в нем есть данные
cards.head()
```

```
# Используем метод merge для объединения двух dataFrame. 
# Чтобы не было переполнения памяти объединим первые 100 строк

data.iloc[:100].merge(cards, on = 'CustomerID', how = 'left')
```

```
# Так как по умолчанию (default) такие параметры как on использую общую колонку для объединения двух dataFrame
# и парметр how - будет left, то их можно не указывать. Результат будет такой же.

data.iloc[:100].merge(cards)
```

```
# Если хотим чтобы объединение было по dataFrame - cards в парметре how используем right

data.iloc[:100].merge(cards, how = 'right')
```

```
# Так же можно объединить два dataFrame с помощью метода .join
# И можно в качестве индекса использовать общую для них колонку - CustomerID

data.iloc[:100].set_index('CustomerID').join(cards.set_index('CustomerID'))
```

### Задача 5

Выведите квантили 0.4 и 0.6 по численным данным новой таблицы там, где это имеет смысл. 

**Решение:**

```
# Актуальные для нас будут столбцы  Количество - Quantity, Цена товара - UnitPrice и Общая сумма  TotalCost
# По другим колонкам квантиль смысла делать нет.

data[['Quantity','UnitPrice','TotalCost']].quantile([0.4 , 0.6])
```

```
# Или для выбора определенных столбцов можно использовать метод .loc

data.loc[:,['Quantity','UnitPrice','TotalCost']].quantile([0.4 , 0.6])
```
