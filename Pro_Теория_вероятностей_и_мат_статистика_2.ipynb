Задача 1

В ящике находится 17 качественных и 6 бракованных деталей. Если деталь качественная, вы зарабатываете 10 рублей. Если бракованная, вы отдаете 20 рублей.

Посчитать среднюю прибыль, если достать две детали наугад.

```
P_2 = 17 / 23 * 16 / 22           # Вероятность выбора двух качественных деталей
P_0 = 6 / 23 * 5 / 22             # Вероятность выбора двух некачественной детали
P_1 = 1 - P_2 - P_0               # Вероятность выбора одной качественной и одной некачественной детали

W = {-40:P_0, -10:P_1, 20:P_2}    # Cоставим закон распределения

sum([W[a]*a for a in W.keys()])   # Математическое ожидание 
```

Задача 2

На испытание поставлено 1200 однотипных ламп.
За первые 2000 часов работы отказало 70 ламп, а за интервал 3000–
4000 часа отказало еще 50 ламп. Определить частоту и интенсивность отказов электронных ламп в промежутке 3000–4000 часов работы.

```
fail_int = 70 / (1200 * 2000)                              # Частота отказов за первые 2000 часов  https://ru.wikipedia.org/wiki/%D0%98%D0%BD%D1%82%D0%B5%D0%BD%D1%81%D0%B8%D0%B2%D0%BD%D0%BE%D1%81%D1%82%D1%8C_%D0%BE%D1%82%D0%BA%D0%B0%D0%B7%D0%BE%D0%B2
print(fail_int)

fail3000 = fail_int * (1200-70) * 1000                      # Кол-во отказавших ламп в период с 2000 до 3000
print(fail3000)

fail_int_4000 = 50 / ((1200 - 70 - fail3000 - 50 + 1200 - 70 - fail3000) / 2 * 1000)       # Интенсивность отказов за период 3000-4000 часов
print(fail_int_4000)

freq_int_4000 = 50 / ((1200 - 70 - fail3000) * 1000)            # Частота отказов за период 3000-4000 часов
print(freq_int_4000)
```

В течение некоторого периода времени производилось наблюдение за работой одного экземпляра радиолокационной станции. За весь период наблюдения было зафиксировано 15 отказов. До начала наблюдения станция проработала
258 часов, а к концу наблюдения наработка станции составила
1233 часа. Требуется определить среднюю наработку на отказ.

Задача 4

Система состоит из пяти приборов, вероятность исправной работы которых в течение времени t = 100 ч
равны: р1(100) = 0,9996; р2(100) = 0,9998; р3(100) = 0,9996;
р4(100) = 0,999; р5(100) = 0,9998. Требуется определить частоту
отказов системы в момент времени t = 100 ч.

```
P1 = 0.9996                                # Вероятность исправной работы 1 прибора 
P2 = 0.9998                                # Вероятность исправной работы 2 прибора 
P3 = 0.9996                                # Вероятность исправной работы 3 прибора 
P4 = 0.999                                 # Вероятность исправной работы 4 прибора 
P5 = 0.9998                                # Вероятность исправной работы 5 прибора 

P = P1 * P2 * P3 * P4 * P5                 # Общая вероятность работы всех приборов в течении 100 часов 

print("Общая вероятность работы всех приборов в течении 100 часов: ", P)

fail_int = (1 - P) / 100                   # Интенсивность отказов за 100 часов всех приборов 

print("Интенсивность отказов за 100 часов всех приборов:           ",fail_int)

fail_freq = fail_int * (1 - fail_int * 100)  # Частота отказов системы в момент времени t = 100 ч.

print("\nОтвет:\nЧастота отказов системы в момент времени t = 100 ч:         ",fail_freq)
```

Задача 5

Мистер Хэ играет в европейскую рулетку по следующей системе: постоянно ставит 250 рублей на «красное». 
Составить закон распределения случайной величины  – его выигрыша. 

Вычислить математическое ожидание выигрыша и округлить его. 

Сколько в среднем проигрывает игрок с каждой ставки?

```
P1 = 18 / 37                      # вероятность выпадания красного
P0 = 1 - P1                       # вероятность невыпадания красного

W = {-250:P0, 250:P1}             # Закон распределения 

E = sum([W[a]*a for a in W.keys()])   # Математическое ожидание

print ("В среднем игрок проигрывает с каждой ставки: ", E)
```
