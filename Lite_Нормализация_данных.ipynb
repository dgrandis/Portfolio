Дано отношение «Автодилеры»:

X5	BMW	2700000	5%

**Преобразуйте отношение к 2НФ**

```python
# Импорт бибилиотек
import pandas as pd
```

```python
# Создаем изначальную таблицу
cars = pd.DataFrame({'Модель':['X5', 'X1', 'Qashqai', 'X-Trail'],   'Фирма':['BMW', 'BMW', 'Nissan', 'Nissan'], 
                     'Цена': [2700000, 1200000,  1100000, 2100000], 'Скидка':['5%', '5%', '5%', '10%']})
cars
```

Для того чтобы данную таблицу преобразовать ко **2ой Нормальной форме** надо разделить на **две** таблицы.

Согласно уроку можно добавить новый столбец **ID**, который будет связывать две эти тоблицы.

- Первая таблица это **ID** и модели
- Вторая таблица **ID** и стоимость

```python
# Создание первой таблицы

# Создание таблицы с ID
cars_id = pd.DataFrame({'ID':[122, 76, 14, 56]})
cars_id
```

```python
cars[['Модель', 'Фирма']]
```

```python
# Создание первой таблицы: связывает ID и название моделей
cars_model = pd.concat([cars_id, cars[['Модель', 'Фирма']]], axis=1)
cars_model
```

```python
# Создание второй таблицы: связывает ID и стоимость
cars_price = pd.concat([cars_id, cars[['Модель', 'Цена', 'Скидка']]], axis=1)
cars_price
```

Теперь можно добавлять новые модели автомобилей в первую таблицу, не зависимо от цены.
