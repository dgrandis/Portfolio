### Задача 3
Оставьте в выборке только те приложения, которые имеют рейтинг в пределах от 5 до 3 (оба значения включительно).

Узнайте число платных и бесплатных приложений в этом списке при помощи методов Pandas.

**Решение:**

```
# Используем df_rating из Задачи 2, чтобы таблица была отсортирована еще и по рейтингу

df_rating_sort = df_rating[(df_rating.Rating >= 3) & (df_rating.Rating <= 5)]       
df_rating_sort
```

```
# Число платных и бесплатных приложений 

# Узнаем какие значения есть в столбце Type, чтобы понимать с чем работать
print ("Значения которые присутствуют в столбце Type:",df_rating_sort.Type.unique())
print ("Общее число значений в столбце Type в выборке (рейтинг от 3 до 5):",df_rating_sort.Type.count())

# Узнаем число бесплатных приложений
df_free = df_rating_sort[df_rating_sort.loc[:,'Type'] == 'Free']
print ("\nЧисло бесплатных приложений (Free):",df_free.Type.count())

# Количество платных
df_paid = df_rating_sort[df_rating_sort.loc[:,'Type'] == 'Paid']
print ("Число бесплатных приложений (Paid):",df_paid.Type.count())

# Проверим действительно сумма бесплатных и платных приложений равняется общему числу приложений из выборки (рейтинг от 3 до 5)
df_rating_sort.Type.count()  == df_paid.Type.count() + df_free.Type.count()

```

### Задача 4

В измененном массиве любым способом установите индексы из целых чисел в порядке возрастания.

**Решение:**

```
# Сделал сортировку по индексам (В массиве из предыдущего задания, там был отсортирован по рейтингу)

# Первый способ 

df_rating_sort_index = df_rating_sort.sort_index()         # Сортируем по индексу и создаем новый dataframe
print (df_rating_sort_index)

# Второй способ

df_rating_sort.sort_index(inplace = True)                  # Сортируем данный dataFrame (inplace = True)
print (df_rating_sort)
```

```
# Если в Задаче 4 нужно было добавить новый столбец с индексами из натуральных чисел в порядке возрастания

import numpy as np                                         # Импортируем библиотеку numpy для создания массива

# Создаем массив из натуральных чисело от 0 до значения которое равно количеству строк 
ind = np.arange(df_rating_sort_index .shape[0])      

# Создаем новый столбец с названием Index и присваеваем ему значение из созданного массива
df_rating_sort_index['Index'] = ind

df_rating_sort_index.head()
# Создался новый столбец Index в dataFrame
```

### Задача 5

Определите процент пустых ячеек к заполненным в начальной таблице. Результат выведите в процентах с округлением до двух знаков. 

**Решение:**

```
# Определим общее количество пропусков
df_null = df.isna().sum().sum()

# Подсчитаем соотношение пропусков к общему количеству элементов
proc = df_null / df.size

print ("Общее количество пропусков в процентах: ",round(100 * proc, 2), '%', sep='')         # С округлением round


# Или можно сделать в одну строку и например использовать для вывода метод format
print ("\nОбщее количество пропусков в процентах: {:.3}% ". format(100 * (df.isna().sum().sum() / df.size)))
```
