```python
from prettytable import PrettyTable

# Создаем таблицу
table = PrettyTable()

# Создаем колонки с названием
table.field_names = ['Column', 'Type']

# Заполняем таблицу
table.add_row(['Фамилия', 'string'])
table.add_row(['Имя', 'string'])
table.add_row(['Возраст', 'int'])
table.add_row(['Дата рождения', 'date'])
table.add_row(['Отдел', 'string'])
table.add_row(['Зарплата', 'int'])
print(table)
```

```python
table = PrettyTable()
table.field_names = ['Column', 'Type']

table.add_row(['Зарплата', 'int'])
table.add_row(['Валюта выплаты', 'string'])
table.add_row(['Дата выплаты', 'date'])
table.add_row(['Сотрудник', 'string'])
print(table)
```

```python
table = PrettyTable()
table.field_names = ['Column', 'Type']

table.add_row(['Название отдела', 'string'])
table.add_row(['Номер комнаты отдела', 'int'])
table.add_row(['Руководитель', 'string'])
table.add_row(['Число сотрудников', 'int'])
table.add_row(['Направление', 'string'])
print(table)
```
