### Задача 1

Используя все возможные известные методы, посчтитайте сумму чисел от 1 до 20 следующим способом:
1. Перебором параметра i в цикле for. 
2. Перебором параметра i в цикле while.
3. При помощи специальной функции для созданного списка.
4. При помощи перебора элементов созданного списка в цикле for. 
5. При помощи перебора элементов созданного списка в цикле while.

**Решение:**

```
#1.

sum = 0
for i in range (1,21):
  sum += i
print (sum)

#2. 

sum = 0 
n = 1
while n <= 20:
  sum += n 
  n += 1
print (sum)

#3.

line =[]
for i in range (1,21):
  line.append(i)      # Дополняем список значениями от 1 до 20

def linesum(line):    # Функция для подсчета суммы в переданном ей списке
  sum = 0
  for i in line:
    sum += i
  return (sum)

print (linesum(line))


#4.

line2 = []                    # Новый список для задачи 4
for i in range (1, 21):
  line2.append(i)

sum = 0

for i in line2:
  sum += i
print (sum)

#5.                       

line3 = []                    # Новый список для задачи 5
for i in range (1, 21):
  line3.append(i)

sum = 0
i = 0

while i < len(line3):
  sum += line3[i]
  i += 1
print (sum)


```

### Задача 2

Используя циклы, напишите код, который создает список (list) путем конкатенации значений данного листа с целыми числами от 1 до (произвольного) n включительно.

Пример:

для списка ["сосиски", "горчица"] при n = 3 результат должен выглядеть так:

['сосиски$\_$1', 'горчица$\_$1', 'сосиски$\_2$', 'горчица$\_$2', 'сосиски$\_$3', 'горчица$\_$3']

```
sample_list = ["мандаринки", "киви", "лимон"]
```

**Решение:**

```
sample_list = ["мандаринки", "киви", "лимон"]

n = int(input("Введите целое положительное число: "))                     #Ввод значения от 1 до (произвольного) n включительно.

if n <= 0:                                                              # Проверка что введенное число положительное и не 0 
  print ("Повторите попытку. Введите целое положительное число")

list_new = []   # Создаем новый список 

for value in range(1,n+1):
  for i in sample_list:
    list_new.append(i + '_' + str(value))
print(list_new)
```

Пример:

для списка [3, "котики", 0.45, 5, {'котики' : 2, 'слоники' : 9}, "слоники", 34] на выходе должны получить число 4

```
list_for_pro_task_2 = [35, 0.24, 3 + 4j, "котики", 0.45, (8, 9), "слоники", {"Мадрид": 3, 'Лондон':5}, 23498]
print(list_for_pro_task_2)
```

**Решение:**

```
list_for_pro_task_2 = [35, 0.24, (3+4j), 'котики', 0.45, (8, 9), 'слоники', {'Мадрид': 3, 'Лондон': 5}, 23498]

n = 0                          # Счетчик количества элементов перед dict
for i in list_for_pro_task_2:
  print (type(i))              # Просто посмотреть типы данных (к задаче не относится)
  if type(i) is dict:
    break
  n += 1
  
print ("В заданном списке до тех пор, пока не встретился элемент типа словарь было", n, "элементов")
```
