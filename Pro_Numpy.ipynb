### Задача 2

* Выполните поэлементное умножение my_array на my_generated_array (раздел "Векторные операции").

* Склейте массивы my_array, my_generated_array в один (могут помочь методы np.concatenate, np.vstack, np.hstack).

* Возьмите склеенный массив и разделите его на равные части (больше двух). Можно сделать вручную через циклы либо воспользоваться методом vsplit.

**Решение:**

```
import numpy as np

multiplication = my_array * my_generated_array      #Поэлементное умножение двух массивов и запись результата в переменную multiplication 
print (multiplication.shape)     #Размер массива после операции поэлементного умножения

c1 = np.vstack((my_array, my_generated_array))    #Объединение двух массивов
c2 = np.concatenate ((my_array, my_generated_array))
print ("Размерность с2 - concatenate:", c2.shape)

c3 = np.hstack((my_array, my_generated_array))
print ("Размерность c3 - hstack", c3.shape)


print ("Размерность c1 - vstack", c1.shape)        #Размер объединенных массивов

cut_array = np.vsplit (c1, 5)      #Разделение массива на 5 равных частей
print ("Длина разрезанного массива", len (cut_array))                     #Длина разрезанного массива
print (cut_array[1].shape)
print (cut_array)

```

### Задача 5

Используя массив из предыдущей задачи, преобразуйте его в новый массив со следующими значениями:

    * "small", если значения меньше 20
    * "medium", если значения в промежутке [20, 30]

**Решение:**

```
signification = three_array.astype(str)                            #Преобразование в строковый тип


signification[three_array < 20] = 'small'                          #Если значение меньше 20, то присвоить 'small'
signification[three_array > 30] = 'large'                          #Если значение больше 30, то присвоить 'large'
signification[(three_array >20) & (three_array <30)] = 'mediun'    #Если значение больше 20 и меньше 30 то присвоить 'medium'

new_string_array =np.array(signification)                          #Создание нового массива с учетом новых значений

print (three_array)
print ("")
print (new_string_array)                                           #Печать нового массива
```
