## **Задача 1**
 
Напишите код, который вычисляет значение следующей математической функции в зависимости от переменной х, где х - вектор
 
$F(x) = \frac{1}{25}AA^Tx + Bx + 5$
 
* Создайте матрицы A и B размера 4 на 4, состоящие из случайных целых чисел в интервале от 1 до 10 (используйте соответствующую функцию модуля np.random)
* Создайте вектор х размера 4, состоящие из случайных целых чисел в интервале от 1 до 5
* Перемножение матриц, и матрицы на вектор должно быть матричное.

````
    # Подключим нужные для работы библиотеки
    import numpy as np
    import matplotlib.pyplot as plt
````

````
# Создадим две матрицы А и В. Формой 4 на 4 заполненные случайными числами с помощью функции random. 
A = np.random.randint(1,10,(4,4))
B = np.random.randint(1,10,(4,4))

# Так же создадим вектор размерностью (4,)
x = np.random.randint(1,5,4)

# Проверим форму матриц и ветора
print ("Размерность матрицы А: ",A.shape)
print ("Размерность матрицы В: ",B.shape)
print ("Размерность вектора х: ",x.shape)

# Напишем формулу, которая будет вычислять значение математической функции.
# Перемножение матриц и ветора на матрицы - матричное. Используем функцию матричного умножения - @ 
# Для матричного умножения матрицы А на саму себя, требуется ее транспонировать - A.T

y = 1/25*((A @ A.T) @ x) + B @ x + 5

# Или можно записать вот так: 

z = ((A @ A.T) @ x)/25 + B @ x + 5  # Что эквивалентно. 


# Стоить заметить, что умножая массив на вектор, на выходе получаем в данном случам горизонтальный вектор.

# Выведем результат функции:

print ("\nх = ", x)
print ("y = ", y)
print ("z = ", z)
print ("\nРазмерность вектора y:",y.shape)
````

````
# С помощью метода linspace создадим множество х - в пределах от -10 до 10 в количестве 10000, чтобы значения синуса были более близки к 0. 
x = np.linspace(-10,10,10000)

# Определим все множество значений 'y' в функции Sin(x)   
y = np.sin(x)

# Создадим переменную s и ей присвоим знак всех значений у (-1 или 1)
s = np.sign(y)

# Создадим список z, который будет хранить порядковый номер всех значений 'х', при котором 'у' будет близким к нулю (не ноль, потому что синус)
z = []
 
# Создаем цикл с количествои итерация меньше на 1, чем кол-во элементов 'уэ(это нужно, чтобы не выйти за предел массива)
for i in range(len(y) - 1):
  if (s[i] and (s[i] != s[i+1])):      # если s[i] == True и текущий знак не равен следующему, тогда       Почему s[i] == True будет написано ниже в другом примере
    z.append(i+1)                      # данный позиционый номер элемента массива добавляется в список z. 

# Создаем полотно побольше
plt.figure(figsize=(10,5))
# Создаем график и передаем значения 'x' и 'y'.
plt.plot(x,y)

# На данном графике отображаем точки красного цвета. Так же передаем значения 'x' и 'y', но только позиционные номера берем из списка z.
plt.plot([x[i] for i in z], [y[i] for i in z],'ro')
plt.grid()     # Для наглядности можно отобразить сетку.
plt.show()
print ("Все позиционные номера в массиве x, при которых у равен нулю: ",z)
````

## **Задача 3** 

Напишите функцию, которая получает на вход пару матриц и проверяет закон коммутативности матричного умножения, то есть свойства $A*B=B*A$, возвращая ```1``` в случае подтверждения и ```0``` в противном случае. Поставьте ```1000``` экспериментов для случайных целочисленных матриц размера ```(5,5)``` элементов, которые принадлежат отрезку ```[-5,5]```, и найдите количество успешных срабатываний.

````
# Создаем функцию
def comm_matrix_mult (matrix1, matrix2):
  '''
      Данная функция проверяет закон коммутитативности матричного умножения.
      Вход:
        - Две матрицы (matrix1 и matrix2)
      Выход: 
        - 1 если матрицы коммутативны
        - 0 если марицы не коммутативны  
  '''
  A = matrix1 @ matrix2
  B = matrix2 @ matrix1
  if (A == B).all():         #  Так же вместо такого сравнения можно использовать такую функцию np.array_equal(A, B):
    return 1
  else: 
    return 0


def matrix_random_5on5():

  ''' 
    Данная функция создает две квадратные матрицы 5 на 5 с произвольными целыми числами от -5 до 5  
  Выход:
    - Считает количество коммутативных матриц в 1000 циклах. И выводит на экран.
  '''

  count = 0
  for i in range (1000):
    A = np.random.randint(-5,6,(5,5))
    B = np.random.randint(-5,6,(5,5))
    count += comm_matrix_mult(A,B)
  return print ("Количество коммутативных матриц: ", count)

# Вызов функции
matrix_random_5on5()

# Видно что матрицы по своей сути не коммутативны. Если поставить 10000 или 100000 результат с большей долей вероятности будет такой же
````

````
# Можно проверить функцию. И создать заведомо коммутативные единичные матрицы

D = np.array([[1,1], [1,1]])
F = np.array([[1,1], [1,1]])

print (D)
print (F)


# Создадим такую же функцию и в качестве параметров передадим единичные матрицы
def m(A, B):
  count = 0
  for i in range (1000):
    count += comm_matrix_mult(A,B)
  return print ("Количество коммутативных матриц: ", count)


# Вызовем данную функцию 
m(D, F)

# Видно что матрицы полностью коммутативны и обе функции работает. 
````
