```
import math
from itertools import takewhile

def is_prime(n):
    return not any(True for i in range(2, math.ceil(math.sqrt(n))) if n % i == 0)

def perfects ():
    i = 2
    while True:
        mersenne = 2**i - 1
        if is_prime(mersenne):
            yield 2**(i-1)*mersenne
        i += 1

if __name__ == '__main__':
    limit = 10000
    for i in takewhile(lambda n: n <= limit, perfects()):
        print (i)
```

```
def prod_three_matrix(matrix1, matrix2, matrix3):

  shape1 = matrix1.shape
  shape2 = matrix2.shape
  shape3 = matrix3.shape

  # Ветвь - 1  (Умножение первой на вторую матриц. И далее возможность умножить на третью. Или третью на результат (1 и 2) )

  # Если можно умножить первую матрицу на вторую 
  if (shape1[1] == shape2[0]):                                  # Смотрим возможность умножения матриц (сраниваем размеры столбцов матрицы 1 и строк матрицы 2)
    intermediate_matrix = np.zeros((shape1[0], shape2[1]))      # Создаем нулевую матрицу с размерами равную кол-ву строк матрицы 1 и столбцов матрицы 2
    for i in range(shape1[0]):
      for j in range(shape2[1]):
        hoarder = 0
        for n in range(shape1[1]):
          hoarder += matrix1[i,n] * matrix2[n,j] 
        intermediate_matrix[i,j] = hoarder
    
    # Если можно умножить результат (1 и 2) на третью матрицу
    if (shape2[1] == shape3[0]):
      total_matrix = np.zeros((shape1[0], shape3[1]))
      for i in range(shape1[0]):
        for j in range(shape3[1]):
          hoarder = 0
          for n in range(shape2[1]):
            hoarder += intermediate_matrix[i,n] * matrix3[n,j] 
          total_matrix[i,j] = hoarder
      return total_matrix
    
    # Если третья матрица такая, что невозмонжо умножить результат (1 и 2) на третью матрицу, но возможно третью умножить на результат (1 и 2). 
    elif (shape3[1] == shape1[0]):
      total_matrix = np.zeros((shape3[0], shape2[1]))
      for i in range(shape3[0]):
        for j in range(shape2[1]):
          hoarder = 0
          for n in range(shape3[1]):
            hoarder += matrix3[i,n] * intermediate_matrix[n,j]
          total_matrix[i,j] = hoarder
      return total_matrix
    
    else:
      return print("Третья матрица не подходящего размера. Умножение невозможно. Результат матричного умножения 1 и 2 матриц:\n", intermediate_matrix)

  
  # Ветвь - 2
  #elif (shape1[1] == shape3[0])

  # Если ни одно из условий не выполнилось и матрицы не подходящего размера, тогда вернуть пустой массив.
  else:
    return np.array([])
  

m = prod_three_matrix(m1, m2, m3)
m
```

```
def whos_first(p1, p2):
    s1 = 0
    s2 = 0
    
    for i in p1:
        if i == 'B':
            break
        s1 += 1
    
    for i in p2:
        if i == 'B':
            break
        s2 += 1
    
    if s1 < s2:
        return "p1"
    elif s2 < s1:
        return "p2"
    else:
        return "tie"
```

```
def solve_hanoi_tower(discs):
    if discs == 0:
        return 0
    elif discs == 1:
        return 1
    elif discs == 2:
        return 2
    elif discs == 3:
        return 7
    s = 7
    for i in range (4, discs+1):
        s = s * 2 + 1
    
    return s


solve_hanoi_tower(4)
```

```
def calc_dice_scores(lst):
    s = 0

    for i in lst:
        s1 = 0 
        for j in i:
            if s1 == j:
                return 0
            s1+= j
  
        s += s1
    return s

lst = [(1,4), (3,4), (2,6)]

calc_dice_scores(lst)
```

```
def calc_dice_scores(lst):
  return sum([a + b for a,b in lst]) if not any([a==b for a, b in lst]) else 0 

lst = [(1,4), (6,6), (2,6)]
calc_dice_scores(lst)
```

# Задача - Мини Судоку

```
def any_duplicates(square):
  lst = []
  for i in square:
    for j in i:
      if j in lst:
        return True
      lst.append(j) 
  return False

square = ([[1,2,3], [4,5,6], [7,8,9]])

any_duplicates(square)
```

# Задача - Палочки

```
pal = int(input("Введите количество палочек: "))
y = 0

while pal>1:
  t = y
  g1 = int(input("Первый игрок берет:"))
  t = pal - g1 
  print (f"На столе остается: {t} палочек")
  if t == 1:
    print ("Победа 1 игрока")
    break
  pal = t

  g2 = int(input("Второй игрок берет: "))
  y = t - g2 
  print (f"На столе остается: {y} палочек")
  if y == 1:
    print ("Победа 2 игрока")
    break
  pal = y
```
