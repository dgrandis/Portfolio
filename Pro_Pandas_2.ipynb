**Решение:**

```
#  Найдем количество уникальных товаров за весь период
print ("Количество уникальных наименований товаров:",data.Description.nunique())

# Найдем страны в в которых были размещены заказы и выведем их алфавитном порядке
country = data.Country.sort_values().unique()                    

print ("\nСтраны в в которых были размещены заказы:",country.tolist())         # Можем передать в качестве списка

print ("\nИли вывести как массив:\n")
country
```

### Задача 5

Найдите количество клиентов, которые оформляли заказ в понедельники. Для этого
* сделайте подвыборку, соответствующую условию "в понедельник",
* посчитайте количество уникальных ID клиентов.

*Для решения этой задачи потребуется использовать параметр индекса weekday (по аналогии с рассмотренными на вебинаре примерами .index.month, .index.day)*.

**Решение:**

```
# С помощью подвыборки найдем все заказы по понедельникам и сохраним в переменную monday
monday = data[data.index.weekday== 0]

# Найдем все уникальные ID клиентов которые заказывали по понедельникам
customer_monday = monday.CustomerID.nunique()

# Выведем результат
print ("Количество клиентов, которые оформляли заказ в понедельники: ",customer_monday)
```

```
# Или можно сделать в одну строчку
print ("Количество клиентов, которые оформляли заказ в понедельники: ",data[data.index.weekday== 0].CustomerID.nunique())
```
